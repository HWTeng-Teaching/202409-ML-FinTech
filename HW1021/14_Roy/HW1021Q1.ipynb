##  Using a little bit of algebra, prove that (4.2) is equivalent to (4.3). In other words, the logistic function representation and logit represen- tation for the logistic regression model are equivalent.

要證明方程 (4.2) 和 (4.3) 等價，首先將 logistic 函數形式重新排列，以導出 logit 表示式。

給定 logistic 函數：

\[
p(X) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 X)}}
\]

我們可以藉由將 \( p(X) \) 用指數式重新排列來推導出 logit 表示式。

### 步驟 1: 將 \( p(X) \) 表示成 \( 1 - p(X) \)

從方程 (4.2) 開始：

\[
p(X) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 X)}}
\]

補集概率 \( 1 - p(X) \) 可以表示為：

\[
1 - p(X) = 1 - \frac{1}{1 + e^{-(\beta_0 + \beta_1 X)}}
\]

### 步驟 2: 化簡 \( 1 - p(X) \)

將 \( 1 - p(X) \) 改寫如下：

\[
1 - p(X) = \frac{(1 + e^{-(\beta_0 + \beta_1 X)}) - 1}{1 + e^{-(\beta_0 + \beta_1 X)}}
\]

\[
1 - p(X) = \frac{e^{-(\beta_0 + \beta_1 X)}}{1 + e^{-(\beta_0 + \beta_1 X)}}
\]

### 步驟 3: 計算比值 \( \frac{p(X)}{1 - p(X)} \)

現在，計算 \( \frac{p(X)}{1 - p(X)} \) 的比值：

\[
\frac{p(X)}{1 - p(X)} = \frac{\frac{1}{1 + e^{-(\beta_0 + \beta_1 X)}}}{\frac{e^{-(\beta_0 + \beta_1 X)}}{1 + e^{-(\beta_0 + \beta_1 X)}}}
\]

由於分母相同，這可以化簡為：

\[
\frac{p(X)}{1 - p(X)} = e^{\beta_0 + \beta_1 X}
\]

### 結論

因此，我們證明了：

\[
\frac{p(X)}{1 - p(X)} = e^{\beta_0 + \beta_1 X}
\]

這與方程 (4.3) 相符，完成了對 logistic 函數形式 (4.2) 和 logit 表示式 (4.3) 等價的證明。
