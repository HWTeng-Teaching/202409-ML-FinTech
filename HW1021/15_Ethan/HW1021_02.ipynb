(a) On average, the fraction of people with an odds of 0.37 of defaulting on their credit card payment can be calculated using the formula:

$$
\frac{p}{1 - p} = 0.37
$$

This gives:

$$
p = \frac{0.37}{1 + 0.37} = 0.27
$$

Thus, the probability is \( p = 0.27 \), or 27%.

---


(b) If an individual has a 16% chance (\( p = 0.16 \)) of defaulting on their credit card payment, the odds of defaulting are calculated as:

$$
\text{odds} = \frac{p}{1 - p} = \frac{0.16}{1 - 0.16} = \frac{0.16}{0.84} = 0.19
$$

So, the odds of defaulting are 0.19.
