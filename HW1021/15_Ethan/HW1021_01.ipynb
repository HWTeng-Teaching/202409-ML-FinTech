The Logistic function is represented as:

$$
p(X) = \frac{e^{\beta_0 + \beta_1 X}}{1 + e^{\beta_0 + \beta_1 X}}
$$

The complementary probability is:

$$
1 - p(X) = \frac{1}{1 + e^{\beta_0 + \beta_1 X}}
$$

Now, taking the ratio of \( p(X) \) and \( 1 - p(X) \):

$$
\frac{p(X)}{1 - p(X)} = e^{\beta_0 + \beta_1 X}
$$

Finally, taking the logarithm of this ratio results in:

$$
\log\left(\frac{p(X)}{1 - p(X)}\right) = \beta_0 + \beta_1 X
$$

This is known as the log-odds or the logit function.
