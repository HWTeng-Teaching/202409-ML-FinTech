### Ch6_Q03

### **(a) As we increase $s$ from 0, the training RSS will: (iv) Steadily decrease.**
#### **Reason:**
- $s=0$ : All coefficients ($\beta_j=0$) are constrained to 0, so the model is completely unfit. The training RSS is at its maximum.
- Training RSS measures the error on the training data.
- As $s$ increases, the constraint becomes less strict, allowing more flexibility in the coefficients. This reduces training RSS because the model fits the training data better.
- A less restrictive constraint (larger $𝑠$) cannot increase the training RSS—it can only remain the same or decrease.


### **(b) As we increase $s$ from 0, the test RSS will: (ii) Decrease initially, and then eventually start increasing in a U shape.**
#### **Reason:**
- Initially, as $s$ increases from 0, the model becomes more flexible and better captures the true relationships, reducing test RSS.
- However, as $s$ becomes very large, the model becomes too flexible, leading to overfitting. Overfitting increases the test RSS as the model starts capturing noise instead of the true signal.


### **(c) As we increase $s$ from 0, the variance will: (iv) Steadily increase.**
#### **Reason:**
- A small $s$ imposes strong constraints, leading to simpler models with low variance.
- As $s$ increases, the constraints loosen, allowing the model to become more flexible. This increases the sensitivity of the model to variations in the training data, resulting in higher variance.


### **(d) As we increase $s$ from 0, the (squared) bias will: (iv) Steadily increase.**
#### **Reason:**
- A small $s$ forces most coefficients to be zero, leading to a highly biased model because it oversimplifies the data.
- As $s$ increases, the model becomes more flexible and better captures the true relationships, reducing bias.
- Bias decreases steadily as $s$ grows.


### **(e) As we increase $s$ from 0, the irreducible error will: (v) Remain constant.**
#### **Reason:**
- The irreducible error is inherent to the data and reflects noise or randomness that cannot be explained by the model, regardless of how flexible or constrained it is.
- Since it does not depend on the model or $s$, it remains constant.


### Summary of Answers:
- **(a)** Training RSS: **(iv) Steadily decrease.**
- **(b)** Test RSS: **(ii) Decrease initially, and then eventually start increasing in a U shape.**
- **(c)** Variance: **(iv) Steadily increase.**
- **(d)** Squared Bias: **(iv) Steadily decrease.**
- **(e)** Irreducible Error: **(v) Remain constant.**
