### Ch6_Q04

### **(a) As we increase $λ$ from 0, the training RSS will:  (iii) Steadily increase.**
#### **Reason:**
- When $λ=0$, the objective is just the residual sum of squares (RSS), and the model fits the data perfectly. As $λ$ increases, the model starts to penalize large coefficients, which reduces the model's ability to fit the training data perfectly.
- **In Ridge regression**, there is no explicit threshold where training RSS first decreases and then increases in an inverted U shape. Instead, as $λ$ increases, the coefficients shrink, and the model will not be able to fit the training data as well. Therefore, training RSS steadily increases as $λ$ increases.


### **(b) As we increase $λ$ from 0, the test RSS will: (ii) Decrease initially, and then eventually start increasing in a U shape.**
#### **Reason:**
- For very small $λ$, the model may overfit the data, resulting in high test RSS due to capturing noise in the training data.
- As $λ$ increases, the model becomes simpler (coefficients shrink), reducing overfitting and improving generalization to the test data, thus lowering the test RSS.
- However, if $λ$ becomes too large, the model becomes too simple (coefficients shrink to zero), leading to underfitting. This increases the test RSS.


### **(c) As we increase $λ$ from 0, the variance will: (v) Remain constant.**
#### **Reason:**
- Variance measures the model's sensitivity to changes in the training data. In Ridge regression, as $λ$ increases, the model coefficients shrink, which typically leads to a more stable model (reduced variance).
- Since Ridge regression is regularized, increasing $λ$ reduces variance by constraining the model's flexibility. However, the variance doesn't exhibit a dramatic change in a consistent direction (it doesn't increase or steadily decrease) because it’s already being controlled by the regularization term. It may remain constant after an initial adjustment.



### **(d) As we increase $λ$ from 0, the (squared) bias will: (iv) Steadily increase.**
#### **Reason:**
- **Bias** refers to the error introduced by approximating a real-world problem with a simplified model. In Ridge regression, as $λ$ increases, the model becomes more constrained (coefficients shrink), which reduces its flexibility and its ability to perfectly fit the training data.
- However, the model becomes less likely to overfit and better generalizes to unseen data. This results in a **decrease in bias** as the model becomes more stable and better captures the underlying patterns.
- Bias decreases steadily because increasing $λ$ improves generalization and reduces model complexity.


### **(e) As we increase $λ$ from 0, the irreducible error will: (v) Remain constant.**
#### **Reason:**
- The **irreducible error** is the noise or randomness inherent in the data that cannot be explained by the model, no matter how sophisticated or regularized the model is.
- Since the irreducible error is independent of the model or the regularization parameter $λ$, it remains constant. It represents the natural variability in the data that cannot be reduced by any model.


### Summary of Answers:
- **(a)** Training RSS: **(iii) Steadily increases.**
- **(b)** Test RSS: **(ii) Decrease initially, and then eventually start increasing in a U shape.**
- **(c)** Variance: **(v) Remain constant.**
- **(d)** Squared Bias: **(iv) Steadily decrease.**
- **(e)** Irreducible Error: **(v) Remain constant.**
