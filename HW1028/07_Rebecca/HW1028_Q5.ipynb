```python
###(a)
def Power():
  print(2 ** 3)

Power()
```

```python
###(b)
def Power2(x,a):
  print(x ** a)

Power2(3, 8)

###(c)
Power2(10, 3)
Power2(8, 17)
Power2(13, 3)
```

```python
###(d)
def Power3(x, a):
  result = x ** a
  return result

result = Power3(3, 8)
print(result)
```

```python
###(e)

import matplotlib.pyplot as plt

def plot_f_x_squared():
    x_values = range(1, 11)
    y_values = [Power3(x, 2) for x in x_values]

    plt.figure()
    plt.plot(x_values, y_values, marker='o')
    plt.xlabel('x')
    plt.ylabel('x^2')
    plt.title('Plot of f(x) = x^2')
    plt.xscale('linear')  # 你可以改為 'log' 以使用對數刻度
    plt.yscale('linear')
    plt.show()

plot_f_x_squared()
```

```python
###(f)

import numpy as np

def PlotPower(x_values, a):
    y_values = [Power3(x, a) for x in x_values]

    plt.figure()
    plt.plot(x_values, y_values, marker='o')
    plt.xlabel('x')
    plt.ylabel(f'x^{a}')
    plt.title(f'Plot of x against x^{a}')
    plt.show()

# 測試範例
PlotPower(np.arange(1, 11), 3)
```
