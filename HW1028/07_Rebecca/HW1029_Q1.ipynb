```python
###This question should be answered using the Weekly data set, which is part of the ISLP package.
###This data is similar in nature to the Smarket data from this chapter’s lab,
###except that it contains 1,089 weekly returns for 21 years, from the beginning of 1990 to the end of 2010.

!pip install ISLP
import ISLP
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

from ISLP import load_data
weekly_data = load_data('Weekly')

weekly_data['Direction'] = (weekly_data['Direction'] == 'Up').astype(int)

train_data = weekly_data[weekly_data['Year'] <= 2008]
test_data = weekly_data[weekly_data['Year'] > 2008]

X_train, y_train = train_data[['Lag2']], train_data['Direction']
X_test, y_test = test_data[['Lag2']], test_data['Direction']


###(e) Repeat (d) using LDA.
###qda模型
qda = QDA()
qda.fit(X_train, y_train)

###在測試集進行預測
y_pred_qda = qda.predict(X_test)

###混合矩陣 正確率
confusion_qda = confusion_matrix(y_test, y_pred_qda)
accuracy_qda = accuracy_score(y_test, y_pred_qda)

print("QDA Confusion Matrix:\n", confusion_qda)
print("QDA Accuracy:", accuracy_qda)

###(f) Repeat (d) using QDA.

# QDA模型
qda = QDA()
qda.fit(X_train, y_train)

# 在測試集上進行預測
y_pred_qda = qda.predict(X_test)

# 計算混淆矩陣和正確率
confusion_qda = confusion_matrix(y_test, y_pred_qda)
accuracy_qda = accuracy_score(y_test, y_pred_qda)

print("QDA Confusion Matrix:\n", confusion_qda)
print("QDA Accuracy:", accuracy_qda)


###(g) Repeat (d) using KNN with K = 1.

# KNN模型 (K=1)
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)

# 在測試集上進行預測
y_pred_knn = knn.predict(X_test)

# 計算混淆矩陣和正確率
confusion_knn = confusion_matrix(y_test, y_pred_knn)
accuracy_knn = accuracy_score(y_test, y_pred_knn)

print("KNN Confusion Matrix (K=1):\n", confusion_knn)
print("KNN Accuracy (K=1):", accuracy_knn)

###(h) Repeat (d) using naive Bayes.

# Naive Bayes模型
nb = GaussianNB()
nb.fit(X_train, y_train)

# 在測試集上進行預測
y_pred_nb = nb.predict(X_test)

# 計算混淆矩陣和正確率
confusion_nb = confusion_matrix(y_test, y_pred_nb)
accuracy_nb = accuracy_score(y_test, y_pred_nb)

print("Naive Bayes Confusion Matrix:\n", confusion_nb)
print("Naive Bayes Accuracy:", accuracy_nb)


###(i) Which of these methods appears to provide the best results on this data?

print("Model Comparison:")
print("QDA Accuracy:", accuracy_qda)
print("KNN Accuracy (K=1):", accuracy_knn)
print("Naive Bayes Accuracy:", accuracy_nb)

###(j) Experiment with different combinations of predictors, includ- ing possible transformations and interactions, for each of the methods. Report the variables, method, and associated confu- sion matrix that appears to provide the best results on the held out data. Note that you should also experiment with values for K in the KNN classifier.

# 使用不同的K值進行KNN測試
best_k = 1
best_accuracy = 0
for k in range(1, 21):  # 測試 K 值從 1 到 20
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred_knn = knn.predict(X_test)
    accuracy_knn = accuracy_score(y_test, y_pred_knn)
    if accuracy_knn > best_accuracy:
        best_accuracy = accuracy_knn
        best_k = k

print(f"Best K for KNN: {best_k} with Accuracy: {best_accuracy}")

```
