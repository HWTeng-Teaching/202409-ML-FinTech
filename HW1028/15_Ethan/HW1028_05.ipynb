```python
import matplotlib.pyplot as plt
import numpy as np

def Power():
    """Prints the result of 2 raised to the power of 3."""
    print(2 ** 3)

def Power2(x, a):
    """Prints x raised to the power of a."""
    print(x ** a)

def Power3(x, a):
    """Returns x raised to the power of a."""
    return x ** a

def plot_function(func, x_values, a):
    """Plots the function defined by func over a range of x_values."""
    y_values = [func(x, a) for x in x_values]
    fig, ax = plt.subplots()
    ax.plot(x_values, y_values, marker='o', label=f'x^{a}')
    ax.set_xlabel("x")
    ax.set_ylabel(f"f(x) = x^{a}")
    ax.set_title(f"Plot of f(x) = x^{a}")
    ax.legend()
    plt.show()

# Part (a)
Power()

# Part (b) - Example usage of Power2
print('The answer to 10^3 is: ')
Power2(10, 3)

print('The answer to 8^17 is: ')
Power2(8, 17)

print('The answer to 131^3 is: ')
Power2(131, 3)

# Part (d) - Example usage of Power3
print('The answer to 10^3 is: ', Power3(10, 3))
print('The answer to 8^17 is: ', Power3(8, 17))
print('The answer to 131^3 is: ', Power3(131, 3))

# Part (e) - Plot f(x) = x^2
x_values = np.arange(1, 11)
plot_function(Power3, x_values, 2)

# Part (f) - Plot for a fixed a value
plot_function(Power3, np.arange(1, 11), 3)
```
