Prove that this is the case. In other words, under the assumption that the observations in the kth class are drawn from a N(µk, σ2) distribution, the Bayes classifer assigns an observation to the class for which the discriminant function is maximized.


{Find k to Maximum}  

\begin{equation}
p_k(x) = \frac{\pi_k \cdot \frac{1}{\sqrt{2 \pi \sigma^2}} \exp\left(-\frac{1}{2 \sigma^2} (x - \mu_k)^2\right)}{\sum_{l=1}^K \pi_l \cdot \frac{1}{\sqrt{2 \pi \sigma^2}} \exp\left(-\frac{1}{2 \sigma^2} (x - \mu_l)^2\right)}
\end{equation}

1.{定義概率密度函數}
- 這個式子表示在給定觀測值 x 的情況下，類別 k 的概率密度函數 p_k(x)。
- π_k 是類別 k 的先驗概率。
- μ_k 是類別 k 的均值。
- σ 是標準差。

---

2.{尋找最大化的 k}
- 為了找出最大化 p_k(x) 的 k，考慮分子：
\begin{equation}
\text{Maximize } \left( \pi_k \cdot \frac{1}{\sqrt{2 \pi \sigma^2}} \exp\left(-\frac{1}{2 \sigma^2} (x - \mu_k)^2\right) \right)
\end{equation}

---

3.{取對數以簡化計算}
- 取對數以方便後續計算：
\begin{equation}
\ln(p_k(x)) = \ln\left( \pi_k \cdot \frac{1}{\sqrt{2 \pi \sigma^2}} \exp\left(-\frac{1}{2 \sigma^2} (x - \mu_k)^2\right) \right) -
              \ln\left( \sum_{l=1}^K \pi_l \cdot \frac{1}{\sqrt{2 \pi \sigma^2}} \exp\left(-\frac{1}{2 \sigma^2} (x - \mu_l)^2\right) \right)
\end{equation}

---

4.{忽略常數項以尋找最大值}
- 忽略與 k 無關的常數項，得：
\begin{equation}
\ln(p_k(x)) \propto \ln(\pi_k) - \frac{1}{2\sigma^2}(x - \mu_k)^2
\end{equation}

---

5.{展開平方項}
- 展開平方項，得到：
\begin{equation}
\ln(\pi_k) - \frac{1}{2\sigma^2} \left( x^2 - 2x\mu_k + \mu_k^2 \right) = -\frac{x^2}{2\sigma^2} + \frac{x\mu_k}{\sigma^2} - \frac{\mu_k^2}{2\sigma^2} + \ln(\pi_k)
\end{equation}

---

6.{獲得判別函數}
- 得到最終的判別函數：
\begin{equation}
\delta_k(x) = \frac{x \mu_k}{\sigma^2} - \frac{\mu_k^2}{2\sigma^2} + \ln(\pi_k)
\end{equation}
