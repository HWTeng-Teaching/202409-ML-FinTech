```python
import numpy as np

# Given coefficients
beta_0 = -6
beta_1 = 0.05
beta_2 = 1

# (a) Calculate the probability
hours_studied = 40
gpa = 3.5

# Logistic function
def logistic_regression_probability(beta_0, beta_1, beta_2, x1, x2):
    linear_combination = beta_0 + beta_1 * x1 + beta_2 * x2
    probability = 1 / (1 + np.exp(-linear_combination))
    return probability

probability_a = logistic_regression_probability(beta_0, beta_1, beta_2, hours_studied, gpa)
print(f"The probability that a student who studies for 40 hours and has an undergrad GPA of 3.5 gets an A in the class is approximately: {probability_a:.4f}")

# (b) How many hours would the student in part (a) need to study to have a 50 % chance of getting an A in the class?
# Solve for x1 when probability is 0.5
def find_hours_for_50_percent(beta_0, beta_1, beta_2, x2):
    # Rearranging the logistic equation for probability = 0.5
    return (np.log(0.5 / (1 - 0.5)) - beta_0 - beta_2 * x2) / beta_1

# Number of hours for (b)
hours_for_50_percent = find_hours_for_50_percent(beta_0, beta_1, beta_2, gpa)
print(f"The student would need to study approximately: {hours_for_50_percent:.2f} hours to have a 50% chance of getting an A in the class.")
```
