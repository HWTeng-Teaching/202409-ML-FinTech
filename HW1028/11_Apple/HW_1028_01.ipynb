```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from statsmodels.api import Logit
from statsmodels.tools import add_constant
```

```python
from ISLP import load_data
df = load_data('Weekly')
print(df)
```

#### Recall (d)

```python
train_data = df[df['Year'] <= 2008]
test_data = df[df['Year'] > 2008]
X_train, y_train = train_data[['Lag2']], train_data['Direction']
X_test, y_test = test_data[['Lag2']], test_data['Direction']
```

#### (e) Repeat (d) using LDA.

```python
lda_model = LinearDiscriminantAnalysis().fit(X_train, y_train)
y_pred_lda = lda_model.predict(X_test)
conf_matrix_e = confusion_matrix(y_test, y_pred_lda)
accuracy_e = accuracy_score(y_test, y_pred_lda)

# Compute the confusion matrix and accuracy
print('Confusion Matrix:')
df_LDA = pd.DataFrame(conf_matrix_e, 
                    index=['Actual Down', 'Actual Up'], 
                    columns=['Predicted Down', 'Predicted Up'])
print(df_LDA)
print(f"LDA_Accuracy: {accuracy_e:.4f}")
```

#### (f) Repeat (d) using QDA.

```python
qda_model = QuadraticDiscriminantAnalysis().fit(X_train, y_train)
y_pred_qda = qda_model.predict(X_test)
conf_matrix_f = confusion_matrix(y_test, y_pred_qda)
accuracy_f = accuracy_score(y_test, y_pred_qda)

print('Confusion Matrix:')
df_QDA = pd.DataFrame(conf_matrix_f, 
                    index=['Actual Down', 'Actual Up'], 
                    columns=['Predicted Down', 'Predicted Up'])
print(df_QDA)
print(f"QDA_Accuracy: {accuracy_f:.4f}")
```

#### (g) Repeat (d) using KNN with K = 1.

```python
knn_model = KNeighborsClassifier(n_neighbors=1).fit(X_train, y_train)
y_pred_knn = knn_model.predict(X_test)
conf_matrix_g = confusion_matrix(y_test, y_pred_knn)
accuracy_g = accuracy_score(y_test, y_pred_knn)

print('Confusion Matrix:')
df_KNN = pd.DataFrame(conf_matrix_g, 
                    index=['Actual Down', 'Actual Up'], 
                    columns=['Predicted Down', 'Predicted Up'])
print(df_KNN)
print(f"KNN with K=1_Accuracy: {accuracy_g:.4f}")
```

#### (h) Repeat (d) using naive Bayes.

```python
nb_model = GaussianNB().fit(X_train, y_train)
y_pred_nb = nb_model.predict(X_test)
conf_matrix_h = confusion_matrix(y_test, y_pred_nb)
accuracy_h = accuracy_score(y_test, y_pred_nb)

print('Confusion Matrix:')
df_Bayes = pd.DataFrame(conf_matrix_h, 
                    index=['Actual Down', 'Actual Up'], 
                    columns=['Predicted Down', 'Predicted Up'])
print(df_Bayes)
print(f"naive Bayes_Accuracy: {accuracy_h:.4f}")
```
