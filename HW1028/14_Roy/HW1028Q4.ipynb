```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ISLP import load_data

# Load the Auto dataset
Auto = load_data('Auto')

# Calculate the median of the 'mpg' column
mpg_median = Auto['mpg'].median()

# Create the binary variable 'mpg01'
Auto['mpg01'] = (Auto['mpg'] > mpg_median).astype(int)

# Display the first few rows of the updated dataset to verify
print(Auto[['mpg', 'mpg01']].head())
```

### (b) Explore the data graphically in order to investigate the associ- ation between mpg01 and the other features. Which of the other features seem most likely to be useful in predicting mpg01? Scat- terplots and boxplots may be useful tools to answer this ques- tion. Describe your findings.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style for better visualization
sns.set(style="whitegrid")

# Create scatter plots for continuous variables against mpg01
features = ['displacement', 'horsepower', 'weight', 'acceleration']
for feature in features:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=Auto[feature], y=Auto['mpg'], hue=Auto['mpg01'], palette="Set1")
    plt.title(f"Scatter Plot of {feature} vs. mpg with mpg01")
    plt.xlabel(feature)
    plt.ylabel('mpg')
    plt.show()

# Create box plots to compare mpg01 distributions across features
for feature in features:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='mpg01', y=feature, data=Auto, palette="Set2")
    plt.title(f"Box Plot of {feature} by mpg01")
    plt.xlabel('mpg01 (0: Low, 1: High)')
    plt.ylabel(feature)
    plt.show()
```

Scatter Plots:

In the scatter plots of displacement, horsepower, and weight against mpg, there is a clear separation in mpg01 (color-coded) values:
Cars with lower values of displacement, horsepower, and weight tend to belong to the mpg01 = 1 (high mileage) group.
Conversely, higher values of these features are associated with mpg01 = 0 (low mileage).
Acceleration does not show a strong pattern with mpg01, indicating it might be less predictive for distinguishing high vs. low mileage.
Box Plots:

The box plots highlight differences in the distributions of displacement, horsepower, and weight for high and low mileage categories.
Cars in the mpg01 = 1 group (high mileage) have significantly lower median values for displacement, horsepower, and weight compared to the mpg01 = 0 group.
Acceleration shows a smaller, less pronounced difference between the mpg01 groups, suggesting it might be a weaker predictor.

```python
from sklearn.model_selection import train_test_split

# 使用已經建立二元變數 'mpg01' 的 Auto 資料集
# Split the data into a training set (70%) and a test set (30%)
train_data, test_data = train_test_split(Auto, test_size=0.3, random_state=42)

# Display the sizes of each dataset to confirm the split
print("Training set size:", train_data.shape[0])
print("Test set size:", test_data.shape[0])
```

### (d) Perform LDA on the training data in order to predict mpg01 using the variables that seemed most associated with mpg01 in (b). What is the test error of the model obtained?

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score

X_train = train_data[features]
y_train = train_data['mpg01']
X_test = test_data[features]
y_test = test_data['mpg01']

# 在訓練資料上進行 LDA 分析
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)

# 在測試集上進行預測
y_pred = lda.predict(X_test)

# 計算測試誤差
test_accuracy = accuracy_score(y_test, y_pred)
test_error = 1 - test_accuracy

print(f"Test Error of LDA Model: {test_error:.4f}")
```


Based on the results, the test error of the LDA model is 0.1610, indicating an accuracy of approximately 83.9%. This suggests that the model is reasonably accurate in predicting whether a car has high or low gas mileage (mpg01), with an error rate of around 16.1%.

###(e) Perform QDA on the training data in order to predict mpg01 using the variables that seemed most associated with mpg01 in (b). What is the test error of the model obtained?

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.metrics import accuracy_score

# Perform QDA on the training data
qda = QuadraticDiscriminantAnalysis()
qda.fit(X_train, y_train)

# Make predictions on the test set
y_pred_qda = qda.predict(X_test)

# Calculate the test error for QDA
test_accuracy_qda = accuracy_score(y_test, y_pred_qda)
test_error_qda = 1 - test_accuracy_qda

print(f"Test Error of QDA Model: {test_error_qda:.4f}")
```

The test error of the QDA model is 0.1695, which indicates an accuracy of approximately 83.05%. This error rate suggests that the QDA model performs similarly to the LDA model, with a test error of around 16.95%, showing that both methods yield comparable results in predicting high or low gas mileage (mpg01).

### (f) Perform logistic regression on the training data in order to pre- dict mpg01 using the variables that seemed most associated with mpg01 in (b). What is the test error of the model obtained?

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Perform logistic regression on the training data
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

# Make predictions on the test set
y_pred_log_reg = log_reg.predict(X_test)

# Calculate the test error for logistic regression
test_accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)
test_error_log_reg = 1 - test_accuracy_log_reg

print(f"Test Error of Logistic Regression Model: {test_error_log_reg:.4f}")
```

The test error of the logistic regression model is 0.1525, which corresponds to an accuracy of approximately 84.75%. This indicates that logistic regression performs similarly to LDA and QDA for predicting high or low gas mileage (mpg01), with a test error rate of around 15.25%.








### (g) Perform naive Bayes on the training data in order to predict mpg01 using the variables that seemed most associated with mpg01 in (b). What is the test error of the model obtained?

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Perform Naive Bayes on the training data
nb = GaussianNB()
nb.fit(X_train, y_train)

# Make predictions on the test set
y_pred_nb = nb.predict(X_test)

# Calculate the test error for Naive Bayes
test_accuracy_nb = accuracy_score(y_test, y_pred_nb)
test_error_nb = 1 - test_accuracy_nb

print(f"Test Error of Naive Bayes Model: {test_error_nb:.4f}")
```

The test error of the Naive Bayes model is 0.1695, which corresponds to an accuracy of approximately 83.05%. This indicates that the Naive Bayes model performs similarly to QDA but with a slightly higher error rate compared to logistic regression and LDA for predicting high or low gas mileage (mpg01). The model's error rate is around 16.95%.

### (h) Perform KNN on the training data, with several values of K, in order to predict mpg01. Use only the variables that seemed most associated with mpg01 in (b). What test errors do you obtain? Which value of K seems to perform the best on this data set?

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Define a range of K values to test
k_values = [1, 3, 5, 7, 9, 11, 13, 15]
test_errors = {}

# Loop over K values and calculate test error for each
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred_knn = knn.predict(X_test)
    test_accuracy_knn = accuracy_score(y_test, y_pred_knn)
    test_errors[k] = 1 - test_accuracy_knn

# Display the test errors for each K
for k, error in test_errors.items():
    print(f"Test Error for K={k}: {error:.4f}")
```

#### KNN Model Test Errors for Different Values of K

Here are the test errors for each value of \( K \) in the KNN model:

- **\( K=1 \)**: Test Error = 0.1525
- **\( K=3 \)**: Test Error = 0.1356
- **\( K=5 \)**: Test Error = 0.1610
- **\( K=7 \)**: Test Error = 0.1695
- **\( K=9 \)**: Test Error = 0.1695
- **\( K=11 \)**: Test Error = 0.1780
- **\( K=13 \)**: Test Error = 0.1780
- **\( K=15 \)**: Test Error = 0.1780

#### Conclusion
Among the tested values, **\( K=3 \)** has the lowest test error (0.1356), suggesting that it performs best on this dataset.
