## 證明最大化 $ p_k(x) $ 等同於最大化判別函數 $ \delta_k(x) $

我們希望證明對於類別 $ k $，最大化 $ p_k(x) $ 等同於最大化 $ \delta_k(x) $。

1. **計算 $ p_k(x) $**

   透過貝氏定理，我們有：
   $$
   p_k(x) = \frac{\pi_k f_k(x)}{\sum_{l=1}^K \pi_l f_l(x)}
   $$

   因為分母對所有類別 $ k $ 相同，所以只需最大化分子：
   $$
   \pi_k f_k(x) = \pi_k \cdot \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x - \mu_k)^2}{2\sigma^2}\right)
   $$

2. **取對數**

   為簡化計算，取對數得到：
   $$
   \log(\pi_k f_k(x)) = \log(\pi_k) - \frac{(x - \mu_k)^2}{2\sigma^2} - \frac{1}{2} \log(2\pi\sigma^2)
   $$

   常數項 $ -\frac{1}{2} \log(2\pi\sigma^2) $ 可以忽略，因此我們要最大化：
   $$
   \log(\pi_k) - \frac{(x - \mu_k)^2}{2\sigma^2}
   $$

3. **展開平方項**

   將 $ (x - \mu_k)^2 $ 展開：
   $$
   = \log(\pi_k) - \frac{x^2}{2\sigma^2} + \frac{x \mu_k}{\sigma^2} - \frac{\mu_k^2}{2\sigma^2}
   $$

   因為 $ -\frac{x^2}{2\sigma^2} $ 與 $ k $ 無關，可以忽略。因此我們有：
   $$
   \delta_k(x) = \frac{x \mu_k}{\sigma^2} - \frac{\mu_k^2}{2\sigma^2} + \log(\pi_k)
   $$

這與判別函數 $ \delta_k(x) $ 的定義相同，因此，最大化 $ p_k(x) $ 等同於最大化 $ \delta_k(x) $。
