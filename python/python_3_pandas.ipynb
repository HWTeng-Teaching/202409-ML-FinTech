### Filter, sort, & Group by

```python
df['b'] # Dataframe of column 'b'
```

```python
df[df['b']>0.5] # Rows where the 'b' column is greater than 5
```

```python
df[ (df['b']> 0.5) & (df['b']<0.9)] # Rows where 0.5 <'b' < 0.9
```

```python
df.sort_values('b') # Sorts values by 'b' in ascending order
```

```python
df.sort_values('b', ascending =False) # Sorts values by 'b' in descending order
```
