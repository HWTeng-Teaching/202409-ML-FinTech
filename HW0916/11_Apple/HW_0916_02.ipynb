```python
!pip install ISLP
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# (a)
# 載入 Boston 數據集
from ISLP import load_data
boston = load_data('Boston')

# 檢查前幾行資料
boston.head()
```

```python
# (b)
# 查看數據的行數和列數
rows, columns = boston.shape
print(f"rows: {rows}, columns: {columns}")
```

行數：數據集中每一行代表一個波士頓郊區（town/suburb）

列數：每一列是不同的變量，如犯罪率、稅率、房屋價格等

```python
# (c)
# 選擇一些變量來繪製散佈圖
sns.pairplot(boston[['crim', 'nox', 'rm', 'age', 'dis']])
plt.show()
```

```python
# (d)
# 計算與犯罪率的相關性
crime_corr = boston.corr()['crim'].sort_values(ascending=False)
print(crime_corr)
```

```python
# (e)
# 查看犯罪率、稅率、師生比的範圍
print(boston[['crim', 'tax', 'ptratio']].describe())

# 查看犯罪率、稅率、師生比的描述性統計
print(boston[['crim', 'tax', 'ptratio']].describe())

# 繪製犯罪率、稅率和師生比的直方圖
boston[['crim', 'tax', 'ptratio']].hist(bins=30, figsize=(12, 8))
plt.tight_layout()
plt.show()
```

Comment:

The crime rate ranges from 0.006 to 88.976. Most suburbs have relatively low crime rates, but there are a few suburbs with particularly high crime rates, indicating that these areas might be less safe.

犯罪率: 範圍廣，從 0.006 到 88.976。大多數郊區的犯罪率較低，但有幾個郊區的犯罪率特別高。這些高犯罪率的郊區可能位於較不安全的社區。
__________________________________________________________
The tax rate ranges from 187 to 711. While the majority of suburbs have moderate tax rates, there are a few that have significantly higher rates. This could be due to the level of public services, infrastructure development, or fiscal policies in those areas.

稅率: 稅率的範圍從 187 到 711。大部分郊區的稅率集中在中低範圍, 但有些郊區的稅率顯著較高, 這可能與當地基礎設施的開發或政府的財政政策有關。
__________________________________________________________
The pupil-teacher ratio ranges from 12.6 to 22.0. Most suburbs have a ratio around 18, but some suburbs have higher ratios, which may indicate fewer educational resources and larger class sizes.

師生比: 師生比的範圍從 12.6 到 22.0。大部分郊區的師生比在中等範圍( 約18 ), 但有些郊區的師生比例明顯較高，這可能意味著當地學校的資源較為有限。
__________________________________________________________
Conclusion: There is considerable variability in crime rates, tax rates, and pupil-teacher ratios across the suburbs of Boston. Some suburbs have particularly high crime rates, while others show high tax rates or pupil-teacher ratios, reflecting the diversity in socio-economic conditions, public services, and education resources.

波士頓郊區的犯罪率、稅率和師生比的範圍都有所不同。一些郊區犯罪率較高，而其他郊區則稅率或師生比較高，這可能反映了當地經濟水平、教育資源和治安狀況的不同。

```python
# (f)
# 計算位於查爾斯河邊的郊區數量
charles_river_count = sum(boston['chas'] == 1)
print(f"The number of the suburbs bound the Charles river: {charles_river_count}")
```

```python
# (g)
# 計算學生與教師比例的中位數
median_ptratio = boston['ptratio'].median()
print(f"The median pupil-teacher ratio: {median_ptratio}")
```

```python
# (i)
# 計算房間數大於7和8的郊區數量
more_than_seven_rooms = sum(boston['rm'] > 7)
more_than_eight_rooms = sum(boston['rm'] > 8)

# 列出房間數大於8的郊區
large_room_suburbs = boston[boston['rm'] > 8]

print(f"The number of the suburbs that average more than seven rooms per dwelling: {more_than_seven_rooms}")
print(f"The number of the suburbs that average more than eight rooms per dwelling: {more_than_eight_rooms}")
print(large_room_suburbs[['rm', 'crim', 'medv']])
```

Result:

Suburbs with more than 7 rooms: There are 64 suburbs where the average number of rooms per dwelling is greater than 7. These suburbs typically represent wealthier areas with larger homes.

有 64 個郊區的平均房間數超過 7, 這些郊區的房屋通常較大，可能代表中高收入的社區。
__________________________________________________________
Suburbs with more than 8 rooms: There are 13 suburbs where the average number of rooms is greater than 8. These suburbs tend to have high median home values (medv) and low crime rates (crim), indicating that they are affluent, safe, and desirable places to live.

有 13 個郊區的平均房間數超過 8, 這些郊區的房價通常較高，且犯罪率(crim)較低，反映這些郊區可能屬於較生活品質較高的社區。
__________________________________________________________
Conclusion: Suburbs with a larger average number of rooms are generally wealthier, with higher home values and lower crime rates. The homes in these areas are spacious, and the communities are likely to have more resources, making them attractive for higher-income families.

房間數較多的郊區通常是波士頓的富裕社區，擁有較低的犯罪率和較高的房價。反映這些郊區的生活品質較高，並且居民可能享有更多的資源和更好的生活環境。
