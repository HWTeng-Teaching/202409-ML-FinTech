```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
```

```python
# (a)
# 讀取文件
path = "C:/Users/apple/Desktop/College.csv"
college = pd.read_csv(path)

# 檢查前幾行資料
college.head()
```

```python
# (b)
# 重命名第一列並設置索引
college2 = pd.read_csv(path, index_col=0)
college3 = college.rename({'Unnamed: 0': 'College'},axis=1)
college3 = college3.set_index('College')

# 查看資料框的列名稱和前幾行資料
college = college3
college.head()
```

```python
# (c)
# 生成數據摘要
college.describe()
```

```python
# (d)
# 選擇幾個變量進行散佈圖矩陣
C = ['Top10perc','Apps','Enroll']
A = pd.DataFrame(college[C])
pd.plotting.scatter_matrix(A)
```

```python
# (e)
college.boxplot('Outstate','Private')
plt.title('Outstate Tuition by Private/Public')
plt.suptitle('')
plt.show()
```

```python
# (f)
# 根據 Top10perc 創建 Elite 變量
college['Elite'] = pd.cut(college['Top10perc'],[0,0.5,1],labels=['No', 'Yes'])

# 查看每個 Elite 類別的數量
print(college['Elite'].value_counts())

# 繪製 Outstate 與 Elite 的箱形圖
college.boxplot(column='Outstate', by='Elite')
plt.title('Outstate Tuition by Elite Status')
plt.suptitle('')
plt.show()
```

```python
# (g)
# 選擇要繪製直方圖的變量
quantitative_vars = ['Apps', 'Accept', 'Enroll', 'Outstate']
# 創建一個 2x2 的繪圖窗口
fig, axes = plt.subplots(2, 2, figsize=(10, 6))
# 展開 axes 陣列，使其可以被迭代
axes = axes.ravel()

# 繪製每個變量的直方圖
for i, var in enumerate(quantitative_vars):
    # 在對應的子圖中繪製直方圖
    axes[i].hist(college[var], bins=(i+1)*10)
    
    # 設置圖的標題和軸標籤
    axes[i].set_title(f'Histogram of {var}')
    axes[i].set_xlabel(var)
    axes[i].set_ylabel('Frequency')

# 調整子圖之間的間距
plt.tight_layout()

# 顯示圖形
plt.show()
```

```python
# (h)
# 計算 Apps 和 Accept 之間的相關性
apps_accept_corr = college['Apps'].corr(college['Accept'])
print(f"申請數和接受數之間的相關性: {apps_accept_corr}")
```

There's a clear positive correlation between the number of applications and the number of acceptances. Schools that receive more applications tend to accept more students.

發現申請數Apps和接受數Accept的相關係數接近 1，表明這兩個變量高度相關。申請數多的學校，通常也接受更多的學生。

```python
# 繪製 Outstate 與 Private 的箱形圖
college.boxplot(column='Outstate', by='Private', figsize=(8, 6))
plt.title('Outstate Tuition by Private/Public')
plt.suptitle('')  # 移除預設子標題
plt.xlabel('Private (Yes=Private, No=Public)')
plt.ylabel('Outstate Tuition')
plt.show()
```

Result: Private colleges generally charge significantly higher out-of-state tuition compared to public colleges, which is evident from the boxplot.

私立大學學費明顯高於公立大學

```python
# 根據 Top10perc 創建 Elite 變量
college['Elite'] = pd.cut(college['Top10perc'], bins=[0, 50, 100], labels=['No', 'Yes'])

# 繪製 Outstate 與 Elite 的箱形圖
college.boxplot(column='Outstate', by='Elite', figsize=(8, 6))
plt.title('Outstate Tuition by Elite Status')
plt.suptitle('')
plt.xlabel('Elite Status')
plt.ylabel('Outstate Tuition')
plt.show()
```

```python
Result: Elite colleges charge higher out-of-state tuition, likely reflecting their prestige and academic standards.

精英學校收取較高的學費：精英學校通常收取更高的學費，這與學校聲譽和學術水準有關。
```
