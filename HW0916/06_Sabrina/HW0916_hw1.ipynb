```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
college_v1 = pd.read_csv("College.csv")
college_v2 = pd.read_csv("College.csv", index_col=0)
college = pd.read_csv("College.csv")
college = college.rename({'Unnamed: 0' : 'College'}, axis = 1)
college
# college_v1
# college_v2
```

```python
college.describe()
```

```python
columns = ['Top10perc', 'Apps', 'Enroll']
pd.plotting.scatter_matrix(college[columns], figsize=(8, 8), diagonal=None, alpha=0.8)
```

```python
college.boxplot(column='Outstate', by='Private', figsize=(8, 6))
plt.title('Outstate Tuition vs Private Institutions')
plt.suptitle('')  # Suppress the automatic subtitle from pandas
plt.xlabel('Institution')
plt.ylabel('Tuition')
plt.show()
```

```python
college['Elite'] = pd.cut(college['Top10perc'],
                          bins=[0, 50, 100],
                          labels=['No', 'Yes'])
elite_counts = college['Elite'].value_counts()
print(elite_counts)
```

```python
college.boxplot(column='Outstate', by='Elite', figsize=(8, 6))

# Set labels and title
plt.title('Outstate Tuition vs Elite Universities')
plt.suptitle('')  # Suppress the automatic subtitle from pandas
plt.xlabel('Elite Status')
plt.ylabel('Tuition')

# Display the plot
plt.show()
```

```python
variables = ['Outstate', 'Top10perc', 'Apps', 'Enroll']
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

college['Outstate'].plot.hist(ax=axes[0, 0], bins=20, color='skyblue', alpha=0.7)
axes[0, 0].set_title('Outstate Tuition (20 bins)')

college['Top10perc'].plot.hist(ax=axes[0, 1], bins=15, color='lightgreen', alpha=0.7)
axes[0, 1].set_title('Top 10% (15 bins)')

college['Apps'].plot.hist(ax=axes[1, 0], bins=25, color='lightcoral', alpha=0.7)
axes[1, 0].set_title('Applications (25 bins)')

college['Enroll'].plot.hist(ax=axes[1, 1], bins=10, color='orange', alpha=0.7)
axes[1, 1].set_title('Enrollment (10 bins)')

plt.tight_layout()
plt.show()
```

Findingsï¼š
1. Tuition of Elite Universities are higher then Outstate Universities.
2. Top 10% university have higher tuition than others.
The higher the tuition, the better ranking a univerisity might be.
