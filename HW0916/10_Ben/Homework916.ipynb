```python
selected_stocks = ['FB', 'GOOG', 'NFLX', 'AMZN', 'AAPL']

for stock in selected_stocks:
    data[stock + '_Return'] = data[stock].pct_change()

returns = data[selected_stocks].mean()
covariance_matrix = data[selected_stocks].cov()

portfolio_means = []
portfolio_stds = []

for _ in range(1000):
    weights = np.random.dirichlet(np.ones(len(selected_stocks)), size=1)
    portfolio_mean = np.dot(weights, returns)[0]
    portfolio_std = np.sqrt(np.dot(weights, np.dot(covariance_matrix, weights.T))[0])
    portfolio_means.append(portfolio_mean)
    portfolio_stds.append(portfolio_std)

plt.figure(figsize=(10, 6))
plt.scatter(portfolio_stds, portfolio_means, color='blue', marker='.', label='Efficiency Frontier')
plt.xlabel('Standard Deviation')
plt.ylabel('Mean')
plt.title('Mean vs. Standard Deviation of Portfolios and Individual Stocks')
plt.grid(True)
plt.scatter(returns_std, returns_mean, color='red', marker='*', label='Individual Stocks')
for i, txt in enumerate(selected_stocks):
    plt.annotate(txt, (returns_std[i], returns_mean[i]), textcoords="offset points", xytext=(0,10), ha='center')
plt.legend()

plt.show()
```
