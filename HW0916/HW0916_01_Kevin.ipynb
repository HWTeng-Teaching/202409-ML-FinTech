```python
#d produce the scatter plotmatrix of ['Top10perc','Apps','Enroll']
column=['Top10perc','Apps','Enroll']
pd.plotting.scatter_matrix(pd.DataFrame(college[column]))
```

```python
#e produce side-by-side boxplots of ('Outside','Private')
college.boxplot('Outside','Private')
```

```python
#f(1) create new variable elite
college['Elite']=pd.cut(college['Top10perc'],[0,0.5,1],labels=['no','yes'])
#f(2) use value_counts() to see how many elite universities there are
college['Elite'].value_counts()
```

```python
#f(3) use boxplot() to produce side-by-side boxplots of outstate vs elite
college.boxplot('Outstate','Elite')
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(2,2)
college['PhD'].hist(ax=ax[0,0],bins=10)
college['Outstate'].hist(ax=ax[0,1],bins=15)
college['Enroll'].hist(ax=ax[1,0],bins=20)
college['Books'].hist(ax=ax[1,1],bins=25)
```

```python
#h what's your discover
plt.scatter(college['F.Undergrad'],college['P.Undergrad'])
plt.xlabel('F.Undergrad')
plt.ylabel('P.Undergrad')
plt.title('F.undergrad vs P.Undergrad')
plt.show()
```
