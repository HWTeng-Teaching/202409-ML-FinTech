```python
import numpy as np

# Create a new 'Elite' column based on 'Top10perc'
college3['Elite'] = np.where(college3['Top10perc'] > 50, 'Yes', 'No')

# Check the first few rows to ensure the new column is created
print(college3[['Top10perc', 'Elite']].head())
```

```python
# Count the number of elite and non-elite universities
elite_counts = college3['Elite'].value_counts()
print(elite_counts)
```

```python
# Create side-by-side boxplots for Outstate tuition by Elite status
college3.boxplot(column='Outstate', by='Elite', figsize=(8, 6))
plt.xlabel('Elite')
plt.ylabel('Outstate Tuition')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

quantitative_vars = ['Apps', 'Accept', 'Enroll', 'Outstate']

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))  # 2 rows, 2 columns of subplots

# Plot histograms with varying bins for each quantitative variable
axs[0, 0].hist(college3['Apps'], bins=30, color='blue')
axs[0, 0].set_title('Apps')

axs[0, 1].hist(college3['Accept'], bins=20, color='blue')
axs[0, 1].set_title('Accept')

axs[1, 0].hist(college3['Enroll'], bins=15, color='blue')
axs[1, 0].set_title('Enroll')

axs[1, 1].hist(college3['Outstate'], bins=25, color='blue')
axs[1, 1].set_title('Outstate')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
