## (g)  Use the plot.hist() method of college to produce some histograms with difering numbers of bins for a few of the quantitative variables.
```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2,2)
college['Enroll'].plot.hist(ax=ax[0,0], bins = 150)
college['Accept'].plot.hist(ax=ax[0,1], bins = 200)
college['Books'].plot.hist(ax=ax[1,0], bins = 10)
college['Grad.Rate'].plot.hist(ax=ax[1,1], bins = 20)
```
