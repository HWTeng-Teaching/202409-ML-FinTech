# Question (f).

Create a new qualitative variable, called **Elite**, by binning the
Top10perc variable into two groups based on whether or not the
proportion of students coming from the top 10% of their high
school classes exceeds 50%.


```python
college['Elite.Rate'] = college['Top10perc'] / college['Enroll']
print(college.head())
```

```python
college["Elite"] = pd.cut(college["Elite.Rate"], bins = [0, 0.5, 1], labels = ["No", "Yes"])
print(college.head())
```

# Question (g).

Use the plot.hist() method of college to produce some histograms with difering numbers of bins for a few of the quantitative variables. The command plt.subplots(2, 2) may be useful: it will divide the plot window into four regions so that four plots can be made simultaneously. By changing the arguments you can divide the screen up in other combinations.

```python
data1 = college['Apps']
data2 = college['Accept']
data3 = college['Enroll']
data4 = college['Grad.Rate']

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].hist(data1, bins=50) # bins = 10
axs[0, 0].set_title('App')

axs[0, 1].hist(data2, bins=50)
axs[0, 1].set_title('Accept')

axs[1, 0].hist(data3, bins=50)
axs[1, 0].set_title('Enroll')

axs[1, 1].hist(data4, bins=50)
axs[1, 1].set_title('Grad.Rate')

plt.tight_layout()
plt.show()
```

# Question (h).
Continue exploring the data, and provide a brief summary of what you discover.

I would like to observe the accept rate and gradutation rate for top100 university.

```python
college['Rank.Rate'] = college['Top25perc'] / college['Enroll']
Top100Uni = college.sort_values('Rank.Rate', ascending=False).head(100)
print(Top100Uni)
```

```python
college['Top25perc_larger_Enroll'] = college['Top25perc'] > college['Enroll']
print(college['Top25perc_larger_Enroll'].value_counts(), end='\n\n')

print('The college which has Top25perc larger than Enroll students.:')
for index, row in college.iterrows():
  if row['Top25perc_larger_Enroll'] == True:
    print(index, row['College'])
```

```python
filtered_college = college[college['Top25perc'] <= college['Enroll']]
Top100Uni = filtered_college.sort_values('Rank.Rate', ascending=False).head(100)
print(Top100Uni)
```

```python
data1 = Top100Uni['Accept'] / Top100Uni['Apps']
data2 = Top100Uni['Grad.Rate']

fig, axs = plt.subplots(2, 1, figsize=(10, 4))

axs[0].hist(data1, bins=50)
axs[0].set_title('Accept')

axs[1].hist(data2, bins=50)
axs[1].set_title('Grad.Rate')

plt.tight_layout()
plt.show()
```
