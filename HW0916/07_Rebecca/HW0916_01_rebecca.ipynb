### (h) Continue exploring the data, and provide a brief summary of what you discover.

```python
# Explore the relationship between 'perc.alumni' and 'Outstate'
plt.scatter(college['perc.alumni'], college['Outstate'])
plt.xlabel('perc.alumni')
plt.ylabel('Outstate')
plt.title('perc.alumni vs. Outstate')
plt.show()

# Calculate the correlation between 'Outstate' and 'F.Undergrad'
corr_outstate_funder = college['Outstate'].corr(college['F.Undergrad'])
print(f"Correlation between Outstate and F.Undergrad: {corr_outstate_funder}")


# Investigate the relationship between 'Private' and 'Grad.Rate'
college.boxplot('Grad.Rate', 'Private')
plt.title('Grad.Rate by Private Status')
plt.show()
```
