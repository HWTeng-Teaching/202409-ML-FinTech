### (a) For each predictor, fit a simple linear regression model to predict the response. Describe your results. In which of the models is there a statistically significant association between the predictor and the response? Create some plots to back up your assertions.

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from ISLP import load_data

# 載入 Boston 資料集
boston = load_data('Boston')

# 儲存結果的資料框
results = []  # Use a list to store results temporarily

# 進行簡單線性回歸
for predictor in boston.columns[:-1]:  # 排除 'crim' 欄位作為預測目標
    X = sm.add_constant(boston[predictor])  # 添加截距項
    y = boston['crim']
    model = sm.OLS(y, X).fit()

    # 儲存每個預測變數的 p 值、係數和 R 平方值
    results.append({  # Append a dictionary to the list
        "Predictor": predictor,
        "P-value": model.pvalues[1],  # 預測變數的 p 值
        "Coefficient": model.params[1],  # 預測變數的係數
        "R-squared": model.rsquared  # R 平方值
    })

# Create the DataFrame after the loop
results = pd.DataFrame(results)

# 篩選出顯著的結果 (p 值 < 0.05)
significant_results = results[results['P-value'] < 0.05].sort_values(by="P-value")

# 顯示顯著預測變數的結果表
print("Significant Predictors for Crime Rate (crim):")
print(significant_results)

# 繪製顯著變數的回歸圖
for predictor in significant_results['Predictor']:
    plt.figure()
    sns.regplot(x=boston[predictor], y=boston['crim'])
    plt.title(f'Relationship between {predictor} and crim')
    plt.xlabel(predictor)
    plt.ylabel('crim')
    plt.show()
```

從圖表可以看到各變數對於crime之關係圖

### (b) Fit a multiple regression model to predict the response using all of the predictors. Describe your results. For which predictors can we reject the null hypothesis H0 : βj = 0?

```python
import pandas as pd
import statsmodels.api as sm
from ISLP import load_data

# 載入 Boston 資料集
boston = load_data('Boston')

# 設定目標變數和所有預測變數
X = sm.add_constant(boston.drop(columns=['crim']))  # 所有預測變數並添加常數項
y = boston['crim']  # 設定犯罪率為目標變數

# 建立多元迴歸模型
multi_model = sm.OLS(y, X).fit()

# 顯示模型摘要
print(multi_model.summary())

# 提取顯著性結果
multi_results = pd.DataFrame({
    "Predictor": X.columns[1:],  # 排除常數項
    "P-value": multi_model.pvalues[1:],  # 排除常數項的 p 值
    "Coefficient": multi_model.params[1:],  # 排除常數項的係數
    "R-squared": multi_model.rsquared  # 模型的總體 R 平方值
})

# 篩選出顯著的預測變數 (p 值 < 0.05)
significant_multi_results = multi_results[multi_results['P-value'] < 0.05].sort_values(by="P-value")

# 顯示顯著預測變數
print("Significant Predictors in Multiple Regression for Crime Rate (crim):")
print(significant_multi_results)
```

 1. 總體模型解釋力
- 模型的 \( R^2 \) 值為 0.449，表示該模型能解釋約 44.9% 的犯罪率變異性。雖然模型有一定的解釋力，但仍存在一些未解釋的變異性。

 2. 顯著性預測變數
根據結果中的 \( p \)-值小於 0.05 的預測變數，我們可以拒絕虛無假設 \( H_0: \beta_j = 0 \) 的變數包括：
   - `rad`: \( p \)-值非常小（接近 0），係數為 0.612，顯示 `rad` 與犯罪率之間有顯著的正相關。
   - `medv`: \( p \)-值為 0.00026，係數為 -0.220，顯示 `medv` 與犯罪率之間有顯著的負相關。
   - `dis`: \( p \)-值為 0.00037，係數為 -1.012，顯示 `dis` 與犯罪率之間有顯著的負相關。
   - `zn`: \( p \)-值為 0.015，係數為 0.046，顯示 `zn` 與犯罪率之間有顯著的正相關。

 3. 其他變數
其餘變數（如 `indus`、`chas`、`nox`、`rm`、`age`、`tax`、`ptratio`、`lstat`）的 \( p \)-值均大於 0.05，因此無法在此模型中顯著地預測犯罪率，這表明它們對犯罪率的影響在本模型中不顯著。

 4. 模型摘要
模型摘要中的條件數（condition number）很大，達到 \( 1.24 \times 10^4 \)，這表明可能存在多重共線性問題。

(d) Is there evidence of non-linear association between any of the predictors and the response? To answer this question, for each predictor X, fit a model of the form
* Y =β0 +β1X+β2X2 +β3X3 +ε.

```python
!pip install ace_tools --user  # Install ace_tools for the current user
import pandas as pd
import numpy as np
import statsmodels.api as sm
# crim	zn	indus	chas	nox	rm	age	dis	rad	tax	ptratio	lstat	medv
# 假設數據已加載到 DataFrame 中，並且 MEDV 是響應變數
data = load_data('Boston')

# 獲取所有預測變數名稱
predictors = ['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'lstat']

# 初始化存放結果的字典
nonlinear_results = {}

# 擬合每個預測變數的模型
for predictor in predictors:
    X = data[predictor]
    X2 = X**2
    X3 = X**3
    X_with_nonlinear = pd.DataFrame({'X': X, 'X2': X2, 'X3': X3})

    # 添加截距項
    X_with_nonlinear = sm.add_constant(X_with_nonlinear)

    # 響應變數
    Y = data['medv']

    # 擬合模型
    model = sm.OLS(Y, X_with_nonlinear).fit()

    # 儲存p值信息
    p_values = model.pvalues
    nonlinear_results[predictor] = {'X^2 p-value': p_values['X2'], 'X^3 p-value': p_values['X3']}

# 將結果轉換為DataFrame以方便查看
results_df = pd.DataFrame(nonlinear_results).T

# If ace_tools is still not found after installation, consider printing the DataFrame:
try:
    import ace_tools as tools
    tools.display_dataframe_to_user(name="Nonlinear Regression Results", dataframe=results_df)
except ModuleNotFoundError:
    print("ace_tools could not be imported. Displaying results_df instead:")
    print(results_df) # Print the DataFrame if ace_tools cannot be imported
```
