 ## In this exercise you will create some simulated data and will fit simple linear regression models to it. Make sure to use the default random number generator with seed set to 1 prior to starting part (a) to ensure consistent results.

### (a) Using the normal() method of your random number generator, create a vector, x, containing 100 observations drawn from a N(0,1) distribution. This represents a feature, X.

```python
import numpy as np

# Set random seed for reproducibility
np.random.seed(1)

# Create a vector x containing 100 observations drawn from N(0, 1) distribution
x = np.random.normal(0, 1, 100)

# Print the first few observations
x
```

### (b) Using the normal() method, create a vector, eps, containing 100 observations drawn from a N(0,0.25) distribution—a normal distribution with mean zero and variance 0.25.

```python
# Create a vector eps containing 100 observations drawn from N(0, 0.25) distribution
eps = np.random.normal(0, 0.25, 100)

# Print the first few observations
eps
```

### (c) Using x and eps, generate a vector y according to the modelY =−1+0.5X+ε. (3.39)What is the length of the vector y? What are the values of β0 and β1 in this linear model?

```python
# Generate a vector y according to the model Y = -1 + 0.5 * X + eps
y = -1 + 0.5 * x + eps

# Print the length of y and the first few observations
print("Length of y:", len(y))
print(y[:10])

# Values of beta0 and beta1 in the linear model
beta0 = -1
beta1 = 0.5
print("beta0:", beta0)
print("beta1:", beta1)
```

###(d) Create a scatterplot displaying the relationship between x and y. Comment on what you observe.

```python
import matplotlib.pyplot as plt

# Create a scatterplot displaying the relationship between x and y
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatterplot of x vs y')
plt.show()
```

### (e) Fit a least squares linear model to predict y using x. Comment on the model obtained. How do βˆ0 and βˆ1 compare to β0 and β1?

```python
import statsmodels.api as sm

# Fit a least squares linear model to predict y using x
x_with_constant = sm.add_constant(x)  # Add a constant term for the intercept
model = sm.OLS(y, x_with_constant).fit()

# Print the model summary
print(model.summary())

# Values of beta0 and beta1 in the linear model
beta0 = -1
beta1 = 0.5
print("beta0:", beta0)
print("beta1:", beta1)

# Comment on the model obtained
beta_hat_0 = model.params[0]
beta_hat_1 = model.params[1]
print("Estimated beta0 (beta_hat_0):", beta_hat_0)
print("Estimated beta1 (beta_hat_1):", beta_hat_1)

# Compare beta_hat_0 and beta_hat_1 to beta0 and beta1
print("Comment: The estimated coefficients beta_hat_0 and beta_hat_1 are close to the true values beta0 and beta1, indicating that the least squares linear model is a good fit for the data.")
```

### (f) Display the least squares line on the scatterplot obtained in (d). Draw the population regression line on the plot, in a different color. Use the legend() method of the axes to create an appro- priate legend.

```python
# Create a scatterplot displaying the relationship between x and y
plt.scatter(x, y, label='Observed Data', color='blue')

# Plot the least squares regression line
plt.plot(x, model.params[0] + model.params[1] * x, color='red', label='Least Squares Line')

# Plot the population regression line
plt.plot(x, beta0 + beta1 * x, color='green', linestyle='--', label='Population Regression Line')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatterplot of x vs y with Regression Lines')
plt.legend()
plt.show()
```

### (g) Now fit a polynomial regression model that predicts y using x and x2. Is there evidence that the quadratic term improves the model fit? Explain your answer.

```python
# Fit a polynomial regression model that predicts y using x and x^2
x_poly = np.column_stack((x, x**2))
x_poly_with_constant = sm.add_constant(x_poly)  # Add a constant term for the intercept
model_poly = sm.OLS(y, x_poly_with_constant).fit()

# Print the model summary for the polynomial regression
print(model_poly.summary())

# Compare the R-squared values to determine if the quadratic term improves the model fit
print("R-squared for linear model:", model.rsquared)
print("R-squared for polynomial model:", model_poly.rsquared)

print("Comment: The R-squared value for the polynomial model can be compared to the linear model's R-squared value to determine if the quadratic term improves the model fit. If the R-squared value for the polynomial model is significantly higher, it suggests that the quadratic term provides a better fit.")
```
