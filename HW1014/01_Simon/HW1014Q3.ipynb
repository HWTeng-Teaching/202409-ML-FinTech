(a) For each predictor, fit a simple linear regression model to predict
the response. Describe your results. In which of the models is
there a statistically significant association between the predictor
and the response? Create some plots to back up your assertions.

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

from ISLP import load_data
Boston = load_data('Boston')

# Example assuming your dataframe is named 'Boston'
# and the column names are 'crim', 'zn', 'indus', etc.

for predictor in Boston.columns:
  if predictor != 'crim':  # Exclude the response variable itself
    model = sm.ols(formula=f'crim ~ {predictor}', data=Boston).fit()
    print(f"Regression model for {predictor}:")
    print(model.summary())

    # Check for statistically significant association (p-value < 0.05)
    if model.pvalues[f'{predictor}'] < 0.05:
      print(f"There is a statistically significant association between {predictor} and crime rate.")

    # Create a scatter plot with regression line
    plt.figure(figsize=(8, 6))
    sns.regplot(x=predictor, y='crim', data=Boston)
    plt.title(f"Relationship between {predictor} and Crime Rate")
    plt.xlabel(predictor)
    plt.ylabel("Crime Rate")
    plt.show()
```

(b) Fit a multiple regression model to predict the response using
all of the predictors. Describe your results. For which predictors
can we reject the null hypothesis $H_0$ : $\beta_j$ = 0?

```python
# Fit the multiple regression model
model_multiple = sm.ols(formula='crim ~ zn + indus + chas + nox + rm + age + dis + rad + tax + ptratio + lstat + medv', data=Boston).fit()

# Print the summary of the model
print(model_multiple.summary())

# Check for statistically significant predictors (p-value < 0.05)
print("\nPredictors for which we can reject the null hypothesis (H0: βj = 0):")
for predictor in model_multiple.pvalues.index:
  if predictor != 'Intercept' and model_multiple.pvalues[predictor] < 0.05:
    print(predictor)
```

(c) How do your results from (a) compare to your results from (b)?
Create a plot displaying the univariate regression coefficients
from (a) on the x-axis, and the multiple regression coefficients
from (b) on the y-axis. That is, each predictor is displayed as a
single point in the plot. Its coefficient in a simple linear regression
model is shown on the x-axis, and its coefficient estimate
in the multiple linear regression model is shown on the y-axis.

```python
# Extract coefficients from simple linear regression models (part a)
univariate_coefficients = {}
for predictor in Boston.columns:
  if predictor != 'crim':
    model = sm.ols(formula=f'crim ~ {predictor}', data=Boston).fit()
    univariate_coefficients[predictor] = model.params[predictor]

# Extract coefficients from multiple linear regression model (part b)
multiple_coefficients = model_multiple.params.drop('Intercept').values

# Create a scatter plot comparing coefficients
plt.figure(figsize=(8, 6))
plt.scatter(univariate_coefficients.values(), multiple_coefficients)
plt.xlabel("Univariate Regression Coefficients")
plt.ylabel("Multiple Regression Coefficients")
plt.title("Comparison of Univariate and Multiple Regression Coefficients")

# Add labels for each point (predictor)
for predictor in univariate_coefficients.keys():
    plt.annotate(predictor, (univariate_coefficients[predictor], multiple_coefficients[list(univariate_coefficients.keys()).index(predictor)]))

plt.show()
```

(d) Is there evidence of non-linear association between any of the
predictors and the response? To answer this question, for each
predictor X, fit a model of the form

$Y = \beta_0 + \beta_1X + \beta_2X^2 + \beta_3X^3 + \epsilon$.

```python
for predictor in Boston.columns:
    if predictor != 'crim':
        formula = f'crim ~ {predictor} + np.power({predictor}, 2) + np.power({predictor}, 3)'
        model_nonlinear = sm.ols(formula=formula, data=Boston).fit()
        print(f"\nNonlinear Regression model for {predictor}:")
        print(model_nonlinear.summary())

        # Check if any of the higher-order terms (X^2, X^3) are significant
        if model_nonlinear.pvalues[f'np.power({predictor}, 2)'] < 0.05 or model_nonlinear.pvalues[f'np.power({predictor}, 3)'] < 0.05:
            print(f"Evidence of non-linear association between {predictor} and crime rate.")
```
