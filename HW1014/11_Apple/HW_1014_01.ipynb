```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

(a) Using the normal( ) method of your random number generator,<br>create a vector, x, containing 100 observations drawn from a <br>N(0, 1) distribution. This represents a feature, X.

```python
np.random.seed(1)
x = np.random.normal(0, 1, 100)
x
```

(b) Using the normal( ) method, create a vector, eps, containing 100<br>
observations drawn from a N(0, 0.25) distribution—a normal<br>
distribution with mean zero and variance 0.25.

```python
eps = np.random.normal(0, np.sqrt(0.25), 100)
eps
```

(c) Using x and eps, generate a vector y according to the model Y = −1 + 0.5X + ε.<br>
What is the length of the vector y?<br> 
What are the values of β₀ and β₁ in this linear model?

```python
y = -1 + 0.5 * x + eps
print(f"Length of y = {len(y)}")  # Answer: 100
print("Values in the linear model:\nβ₀ = -1, β₁ = 0.5")
```

(d) Create a scatterplot displaying the relationship between x and 
y. Comment on what you observe.

```python
plt.scatter(x, y, color='blue', label='Data points')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatterplot of x vs y')
plt.legend()
plt.show()
```

The relationship between 𝑥 and 𝑦 is positive.<br> 
𝑥 increases 0.5 unit, 𝑦 tends to increase 1 unit.
There will be some variation in the data points due to the noise 𝜖 from the 𝑁(0, 0.25) distribution, <br>

(e) Fit a least squares linear model to predict y using x. Comment on the model obtained.<br>
How do β^₀ and β^₁ compare to β₀ and β₁?

```python
# Add constant term for intercept
X = sm.add_constant(x)  
model = sm.OLS(y, X).fit()

# show β^₀ and β^₁
beta_0_hat, beta_1_hat = model.params
df = pd.DataFrame({
    'Coefficient': ['Intercept (β₀)', 'Slope (β₁)'],
    'Estimate': [round(beta_0_hat, 4), round(beta_1_hat, 4)]})
print(df)

# Print the model summary
print(model.summary())
```

(f) Display the least squares line on the scatterplot obtained in (d). <br>
Draw the population regression line on the plot, in a different color. <br>
Use the legend( ) method of the axes to create an appropriate legend.

```python
y_pred = model.predict(X)
y_pop = -1 + 0.5 * x

plt.scatter(x, y, color='blue', label='Data points')

# Plot the least squares line
plt.plot(x, y_pred, color='red', label='Least Squares Line')

# Plot the population regression line
plt.plot(x, y_pop, color='green', linestyle='--', label='Population Line')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatterplot with Least Squares and Population Regression Lines')
plt.legend()  # Add legend to distinguish lines
plt.show()
```

(g) Now fit a polynomial regression model that predicts y using x and x².<br>
Is there evidence that the quadratic term improves the model fit? Explain your answer.

```python
x2 = x ** 2

# Fit the polynomial regression model using x and x^2
X_poly = np.column_stack((x, x2)) # Combine x and x^2
X_poly = sm.add_constant(X_poly)  # Add constant term for intercept

# Fit the polynomial regression model
model_poly = sm.OLS(y, X_poly).fit()

# Print the summary of the polynomial regression model
print(model_poly.summary())

# Compare models
print("\nComparison between linear and polynomial models:")
print(f"Linear model R-squared: {model.rsquared:.4f}")
print(f"Polynomial model R-squared: {model_poly.rsquared:.4f}")

# Check if the quadratic term is significant
p_value_quadratic = model_poly.pvalues[2]
print(f"p-value for the quadratic term (x^2): {p_value_quadratic:.4f}")
```

The R-squared value indicates how well the model explains the variation in 𝑦.<br> 
A higher R-squared value in the polynomial model suggests a better fit.<br>
The p-value for the quadratic term indicates whether 𝑥² is statistically significant. <br>
A low p-value (typically <0.05) means the quadratic term significantly improves the model.

(h) Repeat (a)–(f) after modifying the data generation process in such a way that there is less noise in the data. <br>
The model (3.39) should remain the same. You can do this by decreasing the variance of the normal distribution used to <br>
generate the error term ε in (b). Describe your results.

```python
eps_less_noise = np.random.normal(0, np.sqrt(0.05), 100)  # vs sqrt(0.25)
y_less_noise = -1 + 0.5*x + eps_less_noise

# Fit and plot the model with less noise
less_noise_model = sm.OLS(y_less_noise, X).fit()
y_less_noise_pred = less_noise_model.predict(X)
y_less_noise_pop = -1 + 0.5 * x

plt.scatter(x, y_less_noise, color='blue', label='Data points')

# Plot the least squares line
plt.plot(x, y_less_noise_pred, color='red', label='Least Squares Line')

# Plot the population regression line
plt.plot(x, y_less_noise_pop, color='green', linestyle='--', label='Population Line')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatterplot with Least Squares and Population Regression Lines with less noise')
plt.legend()  
plt.show()
```

(j) What are the confidence intervals for β₀ and β₁ based on the 
original data set,<br>
 the noisier data set, and the less noisy dat 
set? Comment on your results.

```python
origin_conf_int = model.conf_int()
less_noise_conf_int = less_noise_model.conf_int()
more_noise_conf_int = more_noise_model.conf_int()

df_origin = pd.DataFrame(origin_conf_int, columns=['Lower Bound', 'Upper Bound'], index=['β₀', 'β₁'])
df_less_noise = pd.DataFrame(less_noise_conf_int, columns=['Lower Bound', 'Upper Bound'], index=['β₀', 'β₁'])
df_more_noise = pd.DataFrame(more_noise_conf_int, columns=['Lower Bound', 'Upper Bound'], index=['β₀', 'β₁'])

df_origin['Model'] = 'Original'
df_less_noise['Model'] = 'Less Noise'
df_more_noise['Model'] = 'More Noise'

df_combined = pd.concat([df_origin, df_less_noise, df_more_noise])
df_combined = df_combined[['Model', 'Lower Bound', 'Upper Bound']]
df_combined
```

Result: The greater the noise, the wider the confidence interval.
