## In this exercise you will create some simulated data and will fit simple linear regression models to it. Make sure to use the default random number generator with seed set to 1 prior to starting part (a) to ensure consistent results.

## (a) Using the normal() method of your random number generator, create a vector, x, containing 100 observations drawn from a N(0, 1) distribution. This represents a feature, X.

```python
import numpy as np

# Set the seed for reproducibility
np.random.seed(1)

# Generate 100 observations from a N(0, 1) distribution
x = np.random.normal(0, 1, 100)

# Display the generated vector
print(x)
```

## (b) Using the normal() method, create a vector, eps, containing 100 observations drawn from a N(0, 0.25) distribution—a normal distribution with mean zero and variance 0.25.

```python
eps = np.random.normal(0, np.sqrt(0.25), 100)

print(eps)
```

## (c) Using x and eps, generate a vector y according to the model Y = −1+0.5X +e. What is the length of the vector y? What are the values of β0 and β1 in this linear model?

```python
# Generate the vector y according to the model Y = -1 + 0.5X + eps
y = -1 + 0.5 * x + eps

# Display the length of y and a sample of the values
print("Length of y:", len(y))
print("First 10 values of y:", y[:10])
print("β0:", -1)
print("β1:", 0.5)
```

## (d) Create a scatterplot displaying the relationship between x and y. Comment on what you observe.

```python
import matplotlib.pyplot as plt

plt.scatter(x, y, color='blue', alpha=0.7)
plt.title('Scatterplot of X vs Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

## (e)  Fit a least squares linear model to predict y using x. Comment on the model obtained. How do β^0 and βˆ1 compare to β0 and β1?

```python
import statsmodels.api as sm

# Add a constant term to x to include the intercept in the model
X = sm.add_constant(x)

# Fit a linear regression model
model = sm.OLS(y, X).fit()

# Get the estimated coefficients
beta_hat = model.params

# Print the estimated coefficients and model summary
print("Estimated coefficients (β̂0 and β̂1):")
print(f"β^0 (intercept): {beta_hat[0]}")
print(f"βˆ1 (slope): {beta_hat[1]}")
print("\nModel Summary:")
print(model.summary())
```

## (f) Display the least squares line on the scatterplot obtained in (d). Draw the population regression line on the plot, in a diferent color. Use the legend() method of the axes to create an appropriate legend.

```python
# Generate the least squares line
y_hat = beta_hat[0] + beta_hat[1] * x

# Generate the population regression line
y_population = -1 + 0.5 * x

# Create the scatterplot
plt.scatter(x, y, color='blue', alpha=0.7, label='Data points')

# Plot the least squares regression line
plt.plot(x, y_hat, color='red', label='Least squares line')

# Plot the population regression line
plt.plot(x, y_population, color='green', linestyle='dashed', label='Population regression line')

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatterplot with Least Squares and Population Regression Lines')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

## (g) Now fit a polynomial regression model that predicts y using x and x^*2*. Is there evidence that the quadratic term improves the model fit? Explain your answer.

```python
# Create the polynomial term x^2
x_squared = x**2

# Prepare the data with both x and x^2
X_poly = np.column_stack((x, x_squared))

# Add a constant term (intercept) for the model
X_poly = sm.add_constant(X_poly)

# Fit the polynomial regression model
poly_model = sm.OLS(y, X_poly).fit()

# Print the model summary
print("Polynomial Regression Model Summary:")
print(poly_model.summary())
print("\nThere is no strong evidence that the quadratic term (x^2)improves the model fit. The p-value for x^2 is very high, indicating that it is not statistically significant, and the Adjusted R-squared has not improved meaningfully compared to the linear model.")
```

## (h) Repeat (a)–(f) after modifying the data generation process in such a way that there is less noise in the data. The model (3.39) should remain the same. You can do this by decreasing the variance of the normal distribution used to generate the error term e in (b). Describe your results.

```python
eps_lessnoise = np.random.normal(0, np.sqrt(0.05), 100)
y_lessnoise = -1 + 0.5 * x + eps_lessnoise

# Add a constant term to x to include the intercept in the model
X = sm.add_constant(x)

# Fit a linear regression model
model_lessnoise = sm.OLS(y_lessnoise, X).fit()

# Print the estimated coefficients
beta_hat = model.params

# Generate the population regression line
y_population = -1 + 0.5 * x

# Create the scatterplot
plt.scatter(x, y_lessnoise, color='blue', alpha=0.7, label='Data points')

# Plot the least squares regression line
plt.plot(x, model_lessnoise.fittedvalues, color='red', label='Least squares line')

# Plot the population regression line
plt.plot(x, y_population, color='green', linestyle='dashed', label='Population regression line')

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatterplot with Least Squares and Population Regression Lines (Less Noise)')

# Add a legend
plt.legend()
plt.grid(True)
plt.show()
print("The least squares model performs better, closely approximating the true population model. ")
```

## (i) Repeat (a)–(f) after modifying the data generation process in such a way that there is more noise in the data. The model (3.39) should remain the same. You can do this by increasing the variance of the normal distribution used to generate the error term e in (b). Describe your results.

```python
eps_morenoise = np.random.normal(0, np.sqrt(1), 100)
y_morenoise = -1 + 0.5 * x + eps_morenoise

# Add a constant term to x to include the intercept in the model
X = sm.add_constant(x)

# Fit a linear regression model
model_morenoise = sm.OLS(y_morenoise, X).fit()

# Print the estimated coefficients
beta_hat = model.params

# Generate the population regression line
y_population = -1 + 0.5 * x

# Create the scatterplot
plt.scatter(x, y_morenoise, color='blue', alpha=0.7, label='Data points')

# Plot the least squares regression line
plt.plot(x, model_morenoise.fittedvalues, color='red', label='Least squares line')

# Plot the population regression line
plt.plot(x, y_population, color='green', linestyle='dashed', label='Population regression line')

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatterplot with Least Squares and Population Regression Lines (More Noise)')

# Add a legend
plt.legend()
plt.grid(True)
plt.show()
print("The increased noise in the data leads to a poorer model fit, as seen by the divergence between the least squares and population regression lines. ")
```

## (j) What are the confdence intervals for β0 and β1 based on the original data set, the noisier data set, and the less noisy data set? Comment on your results.

```python
conf_interval_original = model.conf_int()
print("Confidence intervals (original data set):")
print(conf_interval_original)

# Get the confidence intervals for the noisier data model
conf_interval_noisier = model_morenoise.conf_int()
print("Confidence intervals (more noise data set):")
print(conf_interval_noisier)

# Get the confidence intervals for the less noisy data model
conf_interval_less_noisy = model_lessnoise.conf_int()
print("Confidence intervals (less noise data set):")
print(conf_interval_less_noisy)
print("Wider confidence interval in more noise data, and narrower confidence interval in less noise data.")
```
