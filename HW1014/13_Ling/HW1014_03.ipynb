## (c) How do your results from (a) compare to your results from (b)? Create a plot displaying the univariate regression coefcients from (a) on the x-axis, and the multiple regression coefcients from (b) on the y-axis. That is, each predictor is displayed as a single point in the plot. Its coefcient in a simple linear regression model is shown on the x-axis, and its coefcient estimate in the multiple linear regression model is shown on the y-axis.

```python
# Dictionary to store simple regression coefficients
simple_regressioncoefficients = {}

# Function to extract coefficients from a simple linear regression model
def get_simple_coefficient(predictor, response="crim"):
    X = sm.add_constant(boston[predictor])
    y = boston[response]
    model = sm.OLS(y, X).fit()
    return model.params[1]  # Return the coefficient for the predictor

for predictor in predictors:
    simple_regressioncoefficients[predictor] = get_simple_coefficient(predictor)

# Prepare for multiple regression
X_multi = sm.add_constant(boston[predictors])  # Add constant for intercept
Y_multi = boston[response]

# Fit the multiple regression model
multiple_model = sm.OLS(Y_multi, X_multi).fit()

# Extract multiple regression coefficients
multiple_regressioncoefficients = multiple_model.params.iloc[1:]  # Exclude the intercept

simple_coefficients = list(simple_regressioncoefficients.values())
multiple_coefficients = multiple_regressioncoefficients.values


# Plotting
plt.figure(figsize=(10, 6))

# Add labels and title
plt.axhline(0, color='red', linestyle='--', linewidth=1)  # Reference line for multiple regression
plt.axvline(0, color='red', linestyle='--', linewidth=1)  # Reference line for simple regression
plt.scatter(simple_coefficients, multiple_coefficients)

plt.xlabel('Simple Regression Coefficients')
plt.ylabel('Multiple Regression Coefficients')
plt.title('Comparison of Simple and Multiple Regression Coefficients')
plt.grid()
plt.show()

```
