```python
import numpy as np
import statsmodels.api as sm

# Step 1: Set random seed
rng = np.random.default_rng(10)

# Step 2: Generate x1, x2, and y
x1 = rng.uniform(0, 1, size=100)
x2 = 0.5 * x1 + rng.normal(size=100) / 10
y = 2 + 2 * x1 + 0.3 * x2 + rng.normal(size=100)

# Step 3: Prepare the data for regression
X = np.column_stack((x1, x2))  # Combine x1 and x2 into a single array
X = sm.add_constant(X)  # Add a constant (intercept) to the model

# Step 4: Fit the linear regression model
model = sm.OLS(y, X).fit()

# Step 5: Get the regression coefficients
coefficients = model.params
print("Regression Coefficients:")
print(f"Intercept (β0): {coefficients[0]}")
print(f"Coefficient for x1 (β1): {coefficients[1]}")
print(f"Coefficient for x2 (β2): {coefficients[2]}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Set random seed and generate x1, x2
rng = np.random.default_rng(10)
x1 = rng.uniform(0, 1, size=100)
x2 = 0.5 * x1 + rng.normal(size=100) / 10

# Step 2: Calculate the correlation between x1 and x2
correlation = np.corrcoef(x1, x2)[0, 1]
print(f"Correlation between x1 and x2: {correlation:.4f}")

# Step 3: Create a scatterplot
plt.figure(figsize=(8, 6))
plt.scatter(x1, x2, alpha=0.7, color='blue')
plt.title("Scatterplot of x1 vs. x2")
plt.xlabel("x1")
plt.ylabel("x2")
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.show()
```

```python
import numpy as np
import statsmodels.api as sm

# Step 1: Set random seed and generate x1, x2, and y
rng = np.random.default_rng(10)
x1 = rng.uniform(0, 1, size=100)
x2 = 0.5 * x1 + rng.normal(size=100) / 10
y = 2 + 2 * x1 + 0.3 * x2 + rng.normal(size=100)

# Step 2: Prepare the data for regression
X = np.column_stack((x1, x2))  # Combine x1 and x2 into a single array
X = sm.add_constant(X)  # Add a constant (intercept) to the model

# Step 3: Fit the linear regression model
model = sm.OLS(y, X).fit()

# Step 4: Get the regression coefficients
coefficients = model.params
print("Estimated Regression Coefficients:")
print(f"Intercept (β̂0): {coefficients[0]:.4f}")
print(f"Coefficient for x1 (β̂1): {coefficients[1]:.4f}")
print(f"Coefficient for x2 (β̂2): {coefficients[2]:.4f}")

# Step 5: Get the summary of the model
print("\nModel Summary:")
print(model.summary())

# Step 6: Hypothesis testing for β1 and β2
p_values = model.pvalues
print(f"\nP-value for β1: {p_values[1]:.4f}")
print(f"P-value for β2: {p_values[2]:.4f}")

# Step 7: Interpretation of results
alpha = 0.05  # Significance level

# Test for null hypothesis H0: β1 = 0
if p_values[1] < alpha:
    print("Reject the null hypothesis H0: β1 = 0 (β1 is significant)")
else:
    print("Fail to reject the null hypothesis H0: β1 = 0 (β1 is not significant)")

# Test for null hypothesis H0: β2 = 0
if p_values[2] < alpha:
    print("Reject the null hypothesis H0: β2 = 0 (β2 is significant)")
else:
    print("Fail to reject the null hypothesis H0: β2 = 0 (β2 is not significant)")
```

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Step 1: Generate original data
rng = np.random.default_rng(10)
x1 = rng.uniform(0, 1, size=100)
x2 = 0.5 * x1 + rng.normal(size=100) / 10
y = 2 + 2 * x1 + 0.3 * x2 + rng.normal(size=100)

# Step 2: Add the mismeasured observation
x1 = np.concatenate([x1, [0.1]])
x2 = np.concatenate([x2, [0.8]])
y = np.concatenate([y, [6]])

# Step 3: Fit the regression models using the new data

# Model (c): Using both x1 and x2
X_both = sm.add_constant(np.column_stack((x1, x2)))
model_both = sm.OLS(y, X_both).fit()
print("\nModel with both x1 and x2:")
print(model_both.summary())

# Model (d): Using only x1
X_single_x1 = sm.add_constant(x1)
model_single_x1 = sm.OLS(y, X_single_x1).fit()
print("\nModel with only x1:")
print(model_single_x1.summary())

# Model (e): Using only x2
X_single_x2 = sm.add_constant(x2)
model_single_x2 = sm.OLS(y, X_single_x2).fit()
print("\nModel with only x2:")
print(model_single_x2.summary())

# Step 4: Analyze the new observation
# Get the leverage of each point
influence = model_both.get_influence()
leverage = influence.hat_matrix_diag

# Identify the index of the new observation
new_observation_index = len(x1) - 1
new_leverage = leverage[new_observation_index]
print(f"\nLeverage of the new observation: {new_leverage:.4f}")

# Calculate Cook's distance for identifying outliers and leverage points
cooks_d = influence.cooks_distance[0]
new_cooks_d = cooks_d[new_observation_index]
print(f"Cook's distance for the new observation: {new_cooks_d:.4f}")

# Plotting leverage vs. residuals
plt.figure(figsize=(10, 6))
plt.scatter(leverage, influence.resid_studentized_internal, label='Observations')
plt.scatter(new_leverage, influence.resid_studentized_internal[new_observation_index], color='red', label='New Observation', s=100)
plt.axhline(y=0, color='black', linestyle='--')
plt.xlabel('Leverage')
plt.ylabel('Studentized Residuals')
plt.title('Leverage vs. Studentized Residuals')
plt.legend()
plt.show()
```
