(e) Fit a least squares linear model to predict y using x. Comment
on the model obtained. How do $\hat{\beta_0}$ and $\hat{\beta_1}$ compare to $\beta_0$ and
$\beta_1$?

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Set seed for reproducibility
np.random.seed(1)

# Generate x and eps
x = np.random.normal(0, 1, 100)
eps = np.random.normal(0, np.sqrt(0.25), 100)

# Generate y according to the model
y = -1 + 0.5 * x + eps

# Reshape x for the linear regression model
x_reshaped = x.reshape(-1, 1)

# Create a linear regression model and fit it to the data
model = LinearRegression()
model.fit(x_reshaped, y)

# Get the estimated coefficients (hat Beta0 and hat Beta1)
hat_Beta0 = model.intercept_
hat_Beta1 = model.coef_[0]

# Display the estimated coefficients
print("Estimated Beta0 (hat_Beta0):", hat_Beta0)
print("Estimated Beta1 (hat_Beta1):", hat_Beta1)

# True Beta0 and Beta1 for comparison
Beta0 = -1
Beta1 = 0.5

print("\nTrue Beta0:", Beta0)
print("True Beta1:", Beta1)
```

Comparison with the true values:

The true $\beta_0$ from the model is -1, and the estimated $\hat{\beta_0}$
  is -0.926. These values are very close, indicating a good estimation of the intercept. The true $\beta_1$ from the model is 0.5, and the estimated $\hat{\beta_1}$ is 0.548. The estimated value is quite close to the true value, showing that the slope coefficient was accurately predicted.

(h) Repeat (a)‚Äì(f) after modifying the data generation process in
such a way that there is less noise in the data. The model (3.39)
should remain the same. You can do this by decreasing the variance
of the normal distribution used to generate the error term
" in (b). Describe your results.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Set a new seed for consistency
np.random.seed(1)

# Generate x from N(0, 1) distribution (as in part a)
x = np.random.normal(0, 1, 100)

# Generate eps with less noise (variance reduced to 0.05)
eps_less_noise = np.random.normal(0, np.sqrt(0.05), 100)

# Generate y according to the model Y = -1 + 0.5X + eps
y_less_noise = -1 + 0.5 * x + eps_less_noise

# Fit a linear regression model
model_less_noise = LinearRegression()
x_reshaped = x.reshape(-1, 1)
model_less_noise.fit(x_reshaped, y_less_noise)

# Get the estimated coefficients
hat_Beta0_less_noise = model_less_noise.intercept_
hat_Beta1_less_noise = model_less_noise.coef_[0]

# Create scatterplot with least squares line and population regression line
plt.figure(figsize=(8, 6))
plt.scatter(x, y_less_noise, color='blue', alpha=0.6, label='Data points with less noise')

# Least squares regression line (fitted model)
y_pred_less_noise = model_less_noise.predict(x_reshaped)
plt.plot(x, y_pred_less_noise, color='red', label='Least Squares Line')

# Population regression line (true model: Y = -1 + 0.5X)
y_true_less_noise = -1 + 0.5 * x
plt.plot(x, y_true_less_noise, color='green', linestyle='--', label='Population Regression Line')

# Add labels, title, and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatterplot with Less Noise, Least Squares and Population Regression Lines')
plt.legend()
plt.grid(True)
plt.show()

# Display the estimated Beta0 and Beta1
print("Estimated Beta0 with less noise:", hat_Beta0_less_noise)
print("Estimated Beta1 with less noise:", hat_Beta1_less_noise)
```

Less Noise in Data: The scatterplot shows that the data points are more tightly clustered around the least squares line, reflecting the reduced variance in the error term. This makes the model's predictions more precise.

Closer Fit: The estimated coefficients $\beta_0$ and $\beta_1$ are even closer to the true values of $\beta_0$ = -1 and $\beta_1$ = 0.5 compared to the original model with more noise.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Set a new seed for consistency
np.random.seed(1)

# Generate x from N(0, 1) distribution (as in part a)
x = np.random.normal(0, 1, 100)

# Generate eps with more noise (variance increased to 1.0)
eps_more_noise = np.random.normal(0, np.sqrt(1.0), 100)

# Generate y according to the model Y = -1 + 0.5X + eps
y_more_noise = -1 + 0.5 * x + eps_more_noise

# Fit a linear regression model
model_more_noise = LinearRegression()
x_reshaped = x.reshape(-1, 1)
model_more_noise.fit(x_reshaped, y_more_noise)

# Get the estimated coefficients
hat_Beta0_more_noise = model_more_noise.intercept_
hat_Beta1_more_noise = model_more_noise.coef_[0]

# Create scatterplot with least squares line and population regression line
plt.figure(figsize=(8, 6))
plt.scatter(x, y_more_noise, color='blue', alpha=0.6, label='Data points with more noise')

# Least squares regression line (fitted model)
y_pred_more_noise = model_more_noise.predict(x_reshaped)
plt.plot(x, y_pred_more_noise, color='red', label='Least Squares Line')

# Population regression line (true model: Y = -1 + 0.5X)
y_true_more_noise = -1 + 0.5 * x
plt.plot(x, y_true_more_noise, color='green', linestyle='--', label='Population Regression Line')

# Add labels, title, and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatterplot with More Noise, Least Squares and Population Regression Lines')
plt.legend()
plt.grid(True)
plt.show()

# Display the estimated Beta0 and Beta1
print("Estimated Beta0 with more noise:", hat_Beta0_more_noise)
print("Estimated Beta1 with more noise:", hat_Beta1_more_noise)
```

(j) What are the confidence intervals for $\beta_0$ and $\beta_1$ based on the
original data set, the noisier data set, and the less noisy data
set? Comment on your results.

```python
import statsmodels.api as sm

# Add a constant to x for intercept calculation (Beta0)
x_with_intercept = sm.add_constant(x_reshaped)

# Fit the OLS model for the original data
model_original = sm.OLS(y, x_with_intercept).fit()

# Fit the OLS model for the less noisy data
model_less_noise = sm.OLS(y_less_noise, x_with_intercept).fit()

# Fit the OLS model for the more noisy data
model_more_noise = sm.OLS(y_more_noise, x_with_intercept).fit()

# Get confidence intervals for Beta0 and Beta1
ci_original = model_original.conf_int()
ci_less_noise = model_less_noise.conf_int()
ci_more_noise = model_more_noise.conf_int()

# Display the confidence intervals
print("Original data set CI:")
print(ci_original)
print("\nLess noisy data set CI:")
print(ci_less_noise)
print("\nMore noisy data set CI:")
print(ci_more_noise)
```

As the noise in the data increases, the confidence intervals widen, indicating greater uncertainty in the estimates of $\beta_0$ and $\beta_1$. Conversely, with less noise, the estimates become more precise, and the confidence intervals shrink.
