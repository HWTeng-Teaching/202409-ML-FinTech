## (a)Use normal() method of random number generator,create a vector , x ,containing 100 observations drawn from a N(0,1) distribution.

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.random.normal(0, 1, 100)
x
```

```python
eps = np.random.normal(0,0.25,100)
eps
```

## (c)Using x and eps,generate a vector y according to the model y=-1+0.5x+ε.What is the length of vector y?What are the value of β0 and β1 in this linear model?

```python
y = -1 + 0.5 * x + eps
y
```

```python
len(y)
```

And β0=-1 and β1=0.5

## (d)Create a scatterplot displaying the relationship between x and y.

```python
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('x vs y')
plt.show()
```

## (e)Fit a least squares linear model to predict y using x.How do (β^)0 and (β^)1 compare to β0 and β1?

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
x_reshape = x.reshape(-1, 1)
model.fit(x_reshape, y)

beta_0 = model.intercept_
beta_1 = model.coef_[0]

beta_0, beta_1
```

It close to the real value of β0 and β1 that β0=-1 and β1=0.5.

## (f)Display the least squares line on scatterplot on the plot which obtained in (d)

```python
plt.scatter(x, y, label='Data')

#Least Squares Line
plt.plot(x, model.predict(x_reshape), color='red', label='Least Squares Line')

#Regression Line
plt.plot(x, -1 + 0.5 * x, color='blue', label='Regression Line')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

## (g)Fit a polynomial regression model that predict y using x and x^2.Is there evidence that the quadratic term improves the model fit?

```python
from sklearn.metrics import mean_squared_error

linear_model = LinearRegression()
linear_model.fit(x_reshape, y)
y_pred_linear = linear_model.predict(x_reshape)

#Feature matrix of x and x^2
x_squared = x ** 2
X_poly = np.column_stack((x, x_squared))


poly_model = LinearRegression()
poly_model.fit(X_poly, y)
y_pred_poly = poly_model.predict(X_poly)


mse_linear = mean_squared_error(y, y_pred_linear)
mse_poly = mean_squared_error(y, y_pred_poly)

mse_linear, mse_poly
```

Because the mean square error of poly. is smaller than linear's.We can say that poly improves the model fit.

## (h)Repeat (a)~(f) after modifying the data generation process in such a way that there's less noise in data.Decreasing the variance of normal distri. used to generate the error term in (b).

```python
#new eps
eps_less_noise = np.random.normal(0, 0.1, 100)

#y
y_less_noise = -1 + 0.5 * x + eps_less_noise


model_less_noise = LinearRegression()
model_less_noise.fit(x_reshape, y_less_noise)

#New coefficient
beta_0_less_noise = model_less_noise.intercept_
beta_1_less_noise = model_less_noise.coef_[0]

plt.scatter(x, y_less_noise, label='Data with less noise')
plt.plot(x, model_less_noise.predict(x_reshape), color='red', label='Least squares line')

plt.plot(x, -1 + 0.5 * x, color='blue', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

beta_0_less_noise, beta_1_less_noise
```

We can see that the coefficients is closer to β0 and β1 respectively.So that is a better linear model too.

## (i)Repeat (a)~(f) after modifying the data generation process in such a way that there's more noise in data.Inreasing the variance of normal distri. used to generate the error term in (b).

```python
#new eps
eps_more_noise = np.random.normal(0, 1, 100)

#y
y_more_noise = -1 + 0.5 * x + eps_more_noise


model_more_noise = LinearRegression()
model_more_noise.fit(x_reshape, y_more_noise)

#new coefficient
beta_0_more_noise = model_more_noise.intercept_
beta_1_more_noise = model_more_noise.coef_[0]

plt.scatter(x, y_more_noise, label='Data with more noise')
plt.plot(x, model_more_noise.predict(x_reshape), color='red', label='Least squares line')

plt.plot(x, -1 + 0.5 * x, color='blue', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

beta_0_more_noise, beta_1_more_noise
```

We can see that the coefficients is far away from β0 and β1 respectively.So that is a worse linear model.

## (j)What are the coincidence interval for β0 and β1 based on original dataset,the noisier dataset,and the less noisy dataset?

```python
import statsmodels.api as sm

#constant
x_with_constant = sm.add_constant(x_reshape)

#original
model_original = sm.OLS(y, x_with_constant).fit()
ci_original = model_original.conf_int()

#less noisy
model_less_noise = sm.OLS(y_less_noise, x_with_constant).fit()
ci_less_noise = model_less_noise.conf_int()

#more noisy
model_more_noise = sm.OLS(y_more_noise, x_with_constant).fit()
ci_more_noise = model_more_noise.conf_int()

ci_original, ci_less_noise, ci_more_noise
```

The narrowist coincidence interval is the less noisy one.The widest one is the more noisy one.
