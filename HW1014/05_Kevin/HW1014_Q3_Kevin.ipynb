## (a) This problem involves the Boston data set.For each predictor, ft a simple linear regression model to predict the response. Describe your results. In which of the models is there a statistically signifcant association between the predictor and the response? Create some plots to back up your assertions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import PolynomialFeatures
import statsmodels.api as sm


file_path = r'C:\Users\\Downloads\boston.csv'  
df = pd.read_csv(file_path)

#'CRIME' is the response variable
y = df['CRIM']  
X_columns = df.columns[:-1]  


results_a = {}
for col in X_columns:
    X = df[[col]]
    X = sm.add_constant(X)  
    model = sm.OLS(y, X).fit()
    results_a[col] = model.summary()


    print(f"Predictor: {col}")
    print(model.summary())
    
 
    plt.figure(figsize=(6,4))
    sns.regplot(x=df[col], y=y, ci=None, line_kws={'color':'red'})
    plt.title(f'Simple Linear Regression: {col} vs CRIME')
    plt.show()
```

Assume alpha=0.5,variables which called CRIM,INDUS,NOX,AGE,RAD,TAX,PTRATIO,LSTAT are have statistically signifcant because their t-value are bigger than 1.96.

## (b) Fit a multiple regression model to predict the response using all of the predictors. Describe your results. For which predictors can we reject the null hypothesis H0 : j = 0?

```python
X = df[X_columns]
X = sm.add_constant(X) 
model_b = sm.OLS(y, X).fit()

print(model_b.summary())
```

Variable "CRIM,ZN,TAX and AGE" reject the null hypothesis since their p-value are smaller than 0.05.

## (c) How do your results from (a) compare to your results from (b)?Create a plot displaying the univariate regression coefcients from (a) on the x-axis, and the multiple regression coefcients from (b) on the y-axis. That is, each predictor is displayed as a single point in the plot. Its coefcient in a simple linear regression model is shown on the x-axis, and its coefcient estimate in the multiple linear regression model is shown on the y-axis

```python
simple_coeffs = []
multiple_coeffs = model_b.params[1:]  #Exclude intercept

for col in X_columns:
    X = df[[col]]
    X = sm.add_constant(X) 
    model = sm.OLS(y, X).fit()
    simple_coeffs.append(model.params[1])  #Store the slope (coefficient) for comparison

plt.figure(figsize=(6,6))
plt.scatter(simple_coeffs, multiple_coeffs)
plt.xlabel('Simple Linear Regression Coefficients')
plt.ylabel('Multiple Linear Regression Coefficients')
plt.title('Comparison of Coefficients')
plt.axhline(0, color='red', linestyle='--')
plt.axvline(0, color='red', linestyle='--')
plt.show()
```

## (d) Is there evidence of non-linear association between any of the predictors and the response? To answer this question, for each predictor X, ft a model of the form Y = 0 + 1X + 2X2 + 3X3 + 

```python
for col in X_columns:
    X = df[[col]]
    poly = PolynomialFeatures(degree=3)
    X_poly = poly.fit_transform(X)
    
    model_poly = sm.OLS(y, X_poly).fit()
    
    print(f"Non-linear Regression for {col}")
    print(model_poly.summary())
    

    X_range = np.linspace(X.min(), X.max(), 100)
    y_poly_pred = model_poly.predict(poly.fit_transform(X_range))
    
    plt.figure(figsize=(6,4))
    plt.scatter(X, y, color='blue')
    plt.plot(X_range, y_poly_pred, color='red', label=f'{col} (degree=3)')
    plt.xlabel(col)
    plt.ylabel('CRIME')
    plt.title(f'Non-linear Regression: {col} vs CRIME')
    plt.show()
```

Except for CRIM,all of the variables has non-linear relationship betwwen CRIME.
