To minimize \( \text{Var}(\alpha X + (1 - \alpha) Y) \), we need to find the value of \( \alpha \) that minimizes this expression. Let:

\[
Z = \alpha X + (1 - \alpha) Y
\]

Then,

\[
\text{Var}(Z) = \text{Var} \left( \alpha X + (1 - \alpha) Y \right)
\]

Expanding this using properties of variance, we get:

\[
= \alpha^2 \text{Var}(X) + (1 - \alpha)^2 \text{Var}(Y) + 2 \alpha (1 - \alpha) \text{Cov}(X, Y)
\]

Substituting the terms for variances and covariance, we can rewrite this as:

\[
= \alpha^2 \sigma_X^2 + (1 - 2\alpha + \alpha^2) \sigma_Y^2 + 2 \alpha (1 - \alpha) \sigma_{XY}
\]

This expands further to:

\[
= (\sigma_X^2 + \sigma_Y^2 - 2 \sigma_{XY}) \alpha^2 + ( -2 \sigma_Y^2 + 2 \sigma_{XY}) \alpha + \sigma_Y^2
\]

To minimize \( \text{Var}(Z) \), take the derivative of \( \text{Var}(Z) \) with respect to \( \alpha \) and set it to zero:

\[
\frac{d}{d \alpha} \text{Var}(Z) = 2 \alpha (\sigma_X^2 + \sigma_Y^2 - 2 \sigma_{XY}) + (-2 \sigma_Y^2 + 2 \sigma_{XY}) = 0
\]

Solving for \( \alpha \):

\[
\Rightarrow \alpha (\sigma_X^2 + \sigma_Y^2 - 2 \sigma_{XY}) = \sigma_Y^2 - \sigma_{XY}
\]

\[
\Rightarrow \alpha = \frac{\sigma_Y^2 - \sigma_{XY}}{\sigma_X^2 + \sigma_Y^2 - 2 \sigma_{XY}}
\]
