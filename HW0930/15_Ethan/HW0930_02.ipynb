```python
#[PCA] ä½¿ç”¨ 'USArrests' æ•¸æ“šï¼šå•é¡Œè§£ç­”èˆ‡ä¸­è‹±æ–‡è§£é‡‹
#(a) ä½¿ç”¨ PCA() å‡½æ•¸æ‰¾å‡ºå‰å››å€‹ä¸»æˆåˆ†è² è·å‘é‡ (Show the first to fourth principal component loadings vectors using PCA() function)

import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm

# å¾ statsmodels ä¸­ç²å– USArrests æ•¸æ“š
# Load the 'USArrests' dataset
df = sm.datasets.get_rdataset('USArrests', 'datasets').data

# æ¨™æº–åŒ–æ•¸æ“š
# Standardize the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

# é€²è¡Œ PCA åˆ†æï¼Œå–å‰å››å€‹ä¸»æˆåˆ†
# Perform PCA with the first four principal components
pca = PCA(n_components=4)
pca.fit(df_scaled)

# é¡¯ç¤ºå‰å››å€‹ä¸»æˆåˆ†çš„è² è·å‘é‡
# Display the first four principal component loadings
loadings = pca.components_
print(loadings)


#ä½¿ç”¨ PCA() å‡½æ•¸ä¾†æå–å‰å››å€‹ä¸»æˆåˆ†è² è·å‘é‡ï¼Œé€™äº›å‘é‡ä»£è¡¨äº†æ•¸æ“šé›†ä¸­æ¯å€‹è®Šé‡å°ä¸»æˆåˆ†çš„è²¢ç»ã€‚
#used the PCA() function to extract the first four principal component loadings,
#which represent the contribution of each variable to the principal components.

#(b) ä½¿ç”¨ np.linalg.eig() è¨ˆç®—å‰å››å€‹ä¸»æˆåˆ†è² è·å‘é‡ (Use np.linalg.eig() to find the first to #fourth principal component loadings vectors)

import numpy as np

# è¨ˆç®—å”æ–¹å·®çŸ©é™£
# Compute the covariance matrix
cov_matrix = np.cov(df_scaled.T)

# ä½¿ç”¨ eig() å‡½æ•¸è¨ˆç®—ç‰¹å¾µå€¼å’Œç‰¹å¾µå‘é‡
# Use eig() function to compute eigenvalues and eigenvectors
eig_values, eig_vectors = np.linalg.eig(cov_matrix)

# é¡¯ç¤ºç‰¹å¾µå‘é‡ï¼ˆä¸»æˆåˆ†è² è·å‘é‡ï¼‰
# Display the eigenvectors (principal component loadings)
print(eig_vectors[:,:])

#ä½¿ç”¨ np.linalg.eig() å‡½æ•¸å¾å”æ–¹å·®çŸ©é™£ä¸­æå–ç‰¹å¾µå€¼å’Œç‰¹å¾µå‘é‡ï¼Œé€™äº›ç‰¹å¾µå‘é‡ä¹Ÿå°±æ˜¯ä¸»æˆåˆ†è² è·å‘é‡ã€‚
#used the np.linalg.eig() function to extract the eigenvalues and eigenvectors from the covariance matrix.
#The eigenvectors are the principal component loadings.

#(c) ä½¿ç”¨ np.linalg.svd() è¨ˆç®—å‰å››å€‹ä¸»æˆåˆ†è² è·å‘é‡
#(Use np.linalg.svd() to find the first to fourth principal component loadings vectors)
# ä½¿ç”¨ SVD åˆ†è§£æ•¸æ“šçŸ©é™£
# Perform Singular Value Decomposition (SVD)
U, S, Vt = np.linalg.svd(df_scaled)

# é¡¯ç¤º V çš„è½‰ç½®ï¼ˆå³ä¸»æˆåˆ†è² è·å‘é‡ï¼‰
# Display V transpose (principal component loadings)
print(Vt[:])

#ä½¿ç”¨ SVD æ–¹æ³•åˆ†è§£æ¨™æº–åŒ–å¾Œçš„æ•¸æ“šçŸ©é™£ï¼Œæˆ‘å€‘å¯ä»¥å¾ V çš„è½‰ç½®ä¸­ç²å¾—ä¸»æˆåˆ†è² è·å‘é‡ã€‚

#ä½¿ç”¨ SVD æ–¹æ³•åˆ†è§£æ¨™æº–åŒ–å¾Œçš„æ•¸æ“šçŸ©é™£ï¼Œæˆ‘å€‘å¯ä»¥å¾ V çš„è½‰ç½®ä¸­ç²å¾—ä¸»æˆåˆ†è² è·å‘é‡ã€‚
#Using Singular Value Decomposition (SVD),
#we decompose the standardized data matrix and extract the principal component loadings from the transpose of ğ‘‰.

#(d) çµæœåˆ†æ (Analysis of Results)
#å•é¡Œï¼šé€™äº›çµæœ (a), (b), å’Œ (c) æ˜¯å¦å®Œå…¨ç›¸åŒï¼Ÿç‚ºä»€éº¼ï¼Ÿ
#(Are the results from (a), (b), and (c) exactly the same? Why or why not?)

#æ˜¯çš„ï¼Œå„˜ç®¡è² è·å‘é‡çš„ç¬¦è™Ÿå’Œæ’åˆ—å¯èƒ½ä¸åŒï¼Œä½†å®ƒå€‘åœ¨æ•¸å­¸ä¸Šæ˜¯ç­‰åƒ¹çš„ã€‚é€™æ˜¯å› ç‚º PCA ä¸­çš„ä¸»æˆåˆ†è² è·å‘é‡å¯ä»¥ç¿»è½‰å…¶ç¬¦è™Ÿï¼Œé€™ä¸æœƒæ”¹è®Šå…¶è§£é‡‹ã€‚
#Yes, although the signs and arrangement of the loadings might differ, they are mathematically equivalent.
#This is because in PCA, principal component loadings can have their signs flipped without affecting the interpretation.
```
