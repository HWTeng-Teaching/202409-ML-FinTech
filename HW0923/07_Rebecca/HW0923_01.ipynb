```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
from scipy.spatial.distance import squareform
```

```python
###(a)step1: 定義不相似度矩陣
dissimilarity_matrix=np.array([
    [0.0, 0.3, 0.4, 0.7],
    [0.3, 0.0, 0.5, 0.8],
    [0.4, 0.5, 0.0, 0.45],
    [0.7, 0.8, 0.45, 0.0]
])
```

```python
###(a)step2:轉換成對應的距離向量格式 linkage way
from scipy.spatial.distance import squareform
distance_vector = squareform(dissimilarity_matrix)
```

```python
###(a)step3:用complete linkage 進行階層聚類  完全連結法 complete linkage
Z = linkage(distance_vector, method = 'complete')
```

```python
##(a)step4:樹狀圖Dendrogram
plt.figure(figsize=(8,4))
dendrogram(Z, labels = ['1', '2','3', '4'])
plt.title("Dendrogram Using Complete Linkage")
plt.xlabel("Observations")
plt.ylabel("Dissimilarity")
plt.show()
```

```python
###(b)step1: 定義不相似
dissimilarity_matrix = np.array([
    [0.0, 0.3, 0.4, 0.7],
    [0.3, 0.0, 0.5, 0.8],
    [0.4, 0.5, 0.0, 0.45],
    [0.7, 0.8, 0.45, 0.0]
])
```

```python
###(b)step2:轉換對應距離向量格式
distance_vector = squareform(dissimilarity_matrix)
```

```python
###(b)step3:用single linkage 階層聚類
Z_single = linkage(distance_vector, method = 'single')
```

```python
###(b)step4:樹狀圖Dendrogram  single linkage（單鏈聚類法）
plt.figure(figsize =(8,4))
dendrogram(Z_single, labels=['1', '2', '3', '4'])
plt.title("Dendrogram Using Single Linkage")
plt.xlabel("Observations")
plt.ylabel("Dissimilarity")
plt.show()
```

```python
###(e) 樹狀圖並進行葉子位置調整
###(e)step1: 定義不相似
dissimilarity_matrix = np.array([
    [0.0, 0.3, 0.4, 0.7],
    [0.3, 0.0, 0.5, 0.8],
    [0.4, 0.5, 0.0, 0.45],
    [0.7, 0.8, 0.45, 0.0]
])
```

```python
###(e)step2:轉換對應距離向量格式
distance_vector = squareform(dissimilarity_matrix)
```

```python
###(e)step3:用single linkage 階層聚類
Z_complete = linkage(distance_vector, method = 'complete')
```

```python
###(e)step4:調整葉子的debdrogram
plt.figure(figsize=(8, 4))
dendrogram(Z_complete, labels=['2', '1', '4', '3'])  # 調整了觀測值順序
plt.title("Equivalent Dendrogram (Leaves Repositioned)")
plt.xlabel("Observations")
plt.ylabel("Dissimilarity")
plt.show()
```
