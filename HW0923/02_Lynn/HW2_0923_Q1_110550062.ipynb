## Question 2.
Suppose that we have four observations, for which we compute a dissimilarity matrix, given by

\begin{bmatrix}
 & 0.3 & 0.4 & 0.7 \\
0.3 &  & 0.5 & 0.8 \\
0.4 & 0.5 &  & 0.45 \\
0.7 & 0.8 & 0.45 &
\end{bmatrix}

For instance, the dissimilarity between the first and second observations is
0.3, and the dissimilarity between the second and fourth observations is 0.8.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

dissimilarity_matrix = np.array([
    [0, 0.3, 0.4, 0.7],
    [0.3, 0, 0.5, 0.8],
    [0.4, 0.5, 0, 0.45],
    [0.7, 0.8, 0.45, 0]
])
```

### (a) On the basis of this dissimilarity matrix, sketch the dendrogram that results from hierarchically clustering these four observations using complete linkage. Be sure to indicate on the plot the height at which each fusion occurs, as well as the observations corresponding to each leaf in the dendrogram.

```python
from scipy.spatial.distance import squareform

condensed_matrix = squareform(dissimilarity_matrix)
Z_complete = linkage(condensed_matrix, method='complete')
plt.figure(figsize=(10, 5))
plt.title('Dendrogram (Complete Linkage)')
plt.xlabel('Observations')
plt.ylabel('Dissimilarity')
dendrogram(Z_complete, labels=['1', '2', '3', '4'])
plt.show()
```

### (b) Repeat (a), this time using single linkage clustering.

```python
Z_single = linkage(condensed_matrix, method="single")
plt.figure(figsize=(10, 5))
plt.title('Dendrogram (Single Linkage)')
plt.xlabel('Observations')
plt.ylabel('Dissimilarity')
dendrogram(Z_single, labels=['1', '2', '3', '4'])
plt.show()
```

### (c) Suppose that we cut the dendrogram obtained in (a) such that two clusters result. Which observations are in each cluster?

Cutting the complete linkage dendrogram (from part a) at height 0.5 results in two clusters:

* Cluster 1: (1, 2)
* Cluster 2: (3, 4)

### (d) Suppose that we cut the dendrogram obtained in (b) such that two clusters result. Which observations are in each cluster?

Cutting the complete linkage dendrogram (from part a) at height 0.45 results in two clusters:

* Cluster 1: (1, 2, 3)
* Cluster 2: (4)

### (e) It is mentioned in this chapter that at each fusion in the den-drogram, the position of the two clusters being fused can be swapped without changing the meaning of the dendrogram. Draw a dendrogram that is equivalent to the dendrogram in (a), for which two or more of the leaves are repositioned, but for which the meaning of the dendrogram is the same.

Swapping the leaves of each fusion at the bottom level can create a new dendrogram that is equivalent to the dendrogram in (a).

```python
plt.figure(figsize=(10, 5))
plt.title('Dendrogram (Complete Linkage)')
plt.xlabel('Observations')
plt.ylabel('Dissimilarity')
dendrogram(Z_complete, labels=['2', '1', '4', '3'])
plt.show()
```
