```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

# 相異矩陣 (dissimilarity matrix)
dissimilarity_matrix = np.array([
    [0, 0.3, 0.4, 0.7],
    [0.3, 0, 0.5, 0.8],
    [0.4, 0.5, 0, 0.45],
    [0.7, 0.8, 0.45, 0]
])

# 使用 complete linkage 進行層次聚類
Z_complete = linkage(dissimilarity_matrix, method='complete')

# 繪製樹狀圖 (complete linkage)
plt.figure(figsize=(8, 4))
dendrogram(Z_complete, labels=['1', '2', '3', '4'])
plt.title('Dendrogram using Complete Linkage')
plt.ylabel('Dissimilarity')
plt.show()

# 使用 single linkage 進行層次聚類
Z_single = linkage(dissimilarity_matrix, method='single')

# 繪製樹狀圖 (single linkage)
plt.figure(figsize=(8, 4))
dendrogram(Z_single, labels=['1', '2', '3', '4'])
plt.title('Dendrogram using Single Linkage')
plt.ylabel('Dissimilarity')
plt.show()

# (c) 剪切 complete linkage 的樹狀圖產生兩個聚類
clusters_complete = fcluster(Z_complete, t=2, criterion='maxclust')
print("Clusters for complete linkage:", clusters_complete)

# (d) 剪切 single linkage 的樹狀圖產生兩個聚類
clusters_single = fcluster(Z_single, t=2, criterion='maxclust')
print("Clusters for single linkage:", clusters_single)

# (e) 樹狀圖重新排列（等效 complete linkage 的 R語法版本）
# 使用 labels=c(2,1,4,3) 重新排序葉子
plt.figure(figsize=(8, 4))
dendrogram(Z_complete, labels=['2', '1', '4', '3'])  # 根據需求重新排序
plt.title('Rearranged Dendrogram (Equivalent to Complete Linkage in R)')
plt.ylabel('Dissimilarity')
plt.show()
```
