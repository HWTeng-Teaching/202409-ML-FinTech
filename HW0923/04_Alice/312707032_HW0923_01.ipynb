# CH12_Q2: Hierarchical Clustering

```python
import numpy as np
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
```

Suppose that we have four observations, for which we compute a dissimilarity matrix, given by

```python
# Dissimilarity matrix from the problem
dissimilarity_matrix = np.array([
    [0, 0.3, 0.4, 0.7],
    [0.3, 0, 0.5, 0.8],
    [0.4, 0.5, 0, 0.45],
    [0.7, 0.8, 0.45, 0]
])
```

For instance, the dissimilarity between the first and second observations is 0.3, and the dissimilarity between the second and fourth observations is 0.8.

### Part (a)

On the basis of this dissimilarity matrix, sketch the dendrogram that results from hierarchically clustering these four observations using complete linkage. Be sure to indicate on the plot the height at which each fusion occurs, as well as the observations corresponding to each leaf in the dendrogram.

```python
# Hierarchical clustering using complete linkage
Z_complete = sch.linkage(sch.distance.squareform(dissimilarity_matrix), method='complete')

# Plot the dendrogram for complete linkage
plt.figure(figsize=(8, 5))
sch.dendrogram(Z_complete, labels=[1, 2, 3, 4], above_threshold_color='blue')
plt.title("Dendrogram - Complete Linkage")
plt.xlabel("Observations")
plt.ylabel("Distance")
plt.show()
```

### Part (b)

Repeat (a), this time using single linkage clustering.

```python
# Hierarchical clustering using single linkage
Z_single = sch.linkage(sch.distance.squareform(dissimilarity_matrix), method='single')

# Plot the dendrogram for single linkage
plt.figure(figsize=(8, 5))
sch.dendrogram(Z_single, labels=[1, 2, 3, 4], above_threshold_color='blue')
plt.title("Dendrogram - Single Linkage")
plt.xlabel("Observations")
plt.ylabel("Distance")
plt.show()
```

### Part (c)

 Suppose that we cut the dendrogram obtained in (a) such that two clusters result. Which observations are in each cluster?

```python
print("Cutting the dendrogram from (a) for two clusters (Complete Linkage):")
print("Cluster 1: {1, 2}")
print("Cluster 2: {3, 4}")
```

### Part (d)

Suppose that we cut the dendrogram obtained in (b) such that two clusters result. Which observations are in each cluster?

```python
print("\nCutting the dendrogram from (b) for two clusters (Single Linkage):")
print("Cluster 1: {1, 2, 3}")
print("Cluster 2: {4}")
```

### Part (e)

It is mentioned in this chapter that at each fusion in the dendrogram, the position of the two clusters being fused can be swapped without changing the meaning of the dendrogram. Draw a dendrogram that is equivalent to the dendrogram in (a), for which two or more of the leaves are repositioned, but for which the meaning of the dendrogram is the same.

```python
# Plot the dendrogram for complete linkage with swapped leaves
plt.figure(figsize=(8, 5))
# Swapping the order of the labels [1, 2, 3, 4] to [2, 1, 4, 3]
sch.dendrogram(Z_complete, labels=[2, 1, 4, 3], above_threshold_color='blue')
plt.title("Dendrogram - Complete Linkage (Swapped Leaves)")
plt.xlabel("Observations")
plt.ylabel("Distance")
plt.show()
```
