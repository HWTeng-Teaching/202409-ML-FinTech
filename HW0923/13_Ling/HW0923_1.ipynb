## (a) On the basis of this dissimilarity matrix, sketch the dendrogramthat results from hierarchically clustering these four observations using complete linkage. Be sure to indicate on the plot theheight at which each fusion occurs, as well as the observations corresponding to each leaf in the dendrogram.

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

dissimilarity_matrix = [
    [0, 0.3, 0.4, 0.7],
    [0.3, 0, 0.5, 0.8],
    [0.4, 0.5, 0, 0.45],
    [0.7, 0.8, 0.45, 0]
]

dist = []
for i in range(len(dissimilarity_matrix)):
    for j in range(i + 1, len(dissimilarity_matrix)):
        dist.append(dissimilarity_matrix[i][j])

link = linkage(dist, method='complete')

plt.figure(figsize=(8, 6))
dendrogram(link, labels=['1', '2', '3', '4'])
plt.title('hierarchical cluster dendrogram(complete linkage)')
plt.xlabel('observation')
plt.ylabel('distance')
plt.show()
```

## (b) Repeat (a), this time using single linkage clustering

```python
link = linkage(dist, method='single')
plt.figure(figsize=(8, 6))
dendrogram(link, labels=['1', '2', '3', '4'])
plt.title('hierarchical cluster dendrogram(single linkage)')
plt.xlabel('observation')
plt.ylabel('distance')
plt.show()
```
