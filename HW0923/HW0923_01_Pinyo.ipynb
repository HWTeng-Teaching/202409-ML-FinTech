```python
# Import the necessary libraries
import numpy as np
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt

# Create the dissimilarity matrix
dissimilarity_matrix = np.array([[0, 0.3, 0.4, 0.7],
                                 [0.3, 0, 0.5, 0.8],
                                 [0.4, 0.5, 0, 0.45],
                                 [0.7, 0.8, 0.45, 0]])

# Convert the dissimilarity matrix into a condensed distance matrix
condensed_matrix = sch.distance.squareform(dissimilarity_matrix)

# Perform hierarchical clustering using complete linkage
Z = sch.linkage(condensed_matrix, method='complete')

# Plot the dendrogram
plt.figure(figsize=(8, 5))
sch.dendrogram(Z, labels=['1', '2', '3', '4'], leaf_rotation=90, leaf_font_size=12)
plt.title('Hierarchical Clustering Dendrogram (Complete Linkage)')
plt.xlabel('Observation')
plt.ylabel('Dissimilarity')
plt.show()
```

```python
# Import the necessary libraries
import numpy as np
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt

# Create the dissimilarity matrix
dissimilarity_matrix = np.array([[0, 0.3, 0.4, 0.7],
                                 [0.3, 0, 0.5, 0.8],
                                 [0.4, 0.5, 0, 0.45],
                                 [0.7, 0.8, 0.45, 0]])

# Convert the dissimilarity matrix into a condensed distance matrix
condensed_matrix = sch.distance.squareform(dissimilarity_matrix)

# Perform hierarchical clustering using single linkage
Z = sch.linkage(condensed_matrix, method='single')

# Plot the dendrogram
plt.figure(figsize=(8, 5))
sch.dendrogram(Z, labels=['1', '2', '3', '4'], leaf_rotation=90, leaf_font_size=12)
plt.title('Hierarchical Clustering Dendrogram (Single Linkage)')
plt.xlabel('Observation')
plt.ylabel('Dissimilarity')
plt.show()
```
