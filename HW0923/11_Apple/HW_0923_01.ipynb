```python
import numpy as np
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt


dissimilarity_matrix = np.array([
    [0.0, 0.3, 0.4, 0.7],
    [0.3, 0.0, 0.5, 0.8],
    [0.4, 0.5, 0.0, 0.45],
    [0.7, 0.8, 0.45, 0.0]
])

# 把不相似度矩陣轉換成壓縮距離矩陣
distance_matrix = sch.distance.squareform(dissimilarity_matrix)
```

```python
# (a) Complete Linkage 完全連結法

plt.figure(figsize=(8, 6))
Z_complete = sch.linkage(distance_matrix, method='complete')
sch.dendrogram(Z_complete, labels=[1, 2, 3, 4])
plt.title('Complete Linkage Dendrogram')
plt.xlabel('Observation')
plt.ylabel('Dissimilarity')
plt.show()
```

```python
# (b) Single Linkage 單一連結法

plt.figure(figsize=(8, 6))
Z_single = sch.linkage(distance_matrix, method='single')
sch.dendrogram(Z_single, labels=[1, 2, 3, 4])
plt.title('Single Linkage Dendrogram')
plt.xlabel('Observation')
plt.ylabel('Dissimilarity')
plt.show()
```

```python
# (e) Complete Linkage 完全連結法

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot the original complete linkage dendrogram (part a)
sch.dendrogram(Z_complete, labels=[1, 2, 3, 4], ax=ax1)
ax1.set_title('Original Complete Linkage Dendrogram')
ax1.set_xlabel('Observation')
ax1.set_ylabel('Dissimilarity')

# Plot the rearranged complete linkage dendrogram (swap leaf order to [3, 4, 1, 2])
# 繪製重新排列葉子的完全連結樹狀圖
sch.dendrogram(Z_complete, labels=[2, 1, 4, 3], ax=ax2)
ax2.set_title('Rearranged Complete Linkage Dendrogram')
ax2.set_xlabel('Observation')
ax2.set_ylabel('Dissimilarity')

plt.tight_layout()
plt.show()
```

The order of the leaves is rearranged to 3,4,1,2, effectively swapping the two clusters.

Even though the leaf order is changed, the hierarchical structure and clustering process (the merges and heights) remain the same, showing that the dendrogram’s meaning hasn’t changed.
